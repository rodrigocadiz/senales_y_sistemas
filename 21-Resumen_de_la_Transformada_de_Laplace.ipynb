## Propiedades de la Transformada unilateral de Laplace

<div class="alert alert-block alert-info">
Para este resumen consideramos las siguientes transformadas de señales causales:
$$ f(t) \rightarrow F(s)  \;\;\;\;\; \mbox{ }\;\;\;\;\; g(t) \rightarrow G(s) \;\;\;\;\; \mbox{y}\;\;\;\;\; h(t) \rightarrow H(s) $$

son sus respectivas regiones de converegencia:

$$ \alpha_f < \sigma \;\;\;\;\; \mbox{ }\;\;\;\;\; \alpha_g < \sigma  \;\;\;\;\; \mbox{ }\;\;\;\;\; \alpha_h < \sigma $$
</div>

***
### Linealidad
$$ \boxed{af(t)+bg(t) \rightarrow aF(s)+ bG(s)} \;\;\;\;\; \mbox{ROC:}\;\; \max(\alpha_f,\alpha_g) < \sigma $$


### Desplazamiento

Para todo $a \in \mathbb{R}$

\begin{equation}
\boxed{f(t-a) \rightarrow e^{-a s} F(s)}  \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma 
\end{equation}

\begin{equation}
\boxed{e^{a t}f(t) \rightarrow  F(s-a)} \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma-a
\end{equation}


### Escalamiento
Para $a \in \mathbb{R}/\left\{0\right\}$

\begin{equation}
\boxed{f(at) \rightarrow  \frac{1}{|a|}F(\frac{s}{a})} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \frac{\sigma}{a}  
\end{equation}

### Derivada
- **Derivación en frecuencia**. 
$$ \boxed{t f(t) \rightarrow -F'(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma $$ 

$$ \boxed{t^{n} f(t) \rightarrow (-1)^nF^{(n)}(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma  $$ 

- **Derivación en el tiempo o espacio**. 
$$ \boxed{f'(t) \rightarrow s F(s) -f(0^{+})} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma  $$

$$ \boxed{f^{(n)}(t) \rightarrow s^n F(s) -f^{(n-1)}(0^{+}) -sf^{n-2}(0^{+}) - \dots - s^{n-1}f(0^{+}))} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma   $$

### Convolución
$$\boxed{(h \ast f)(s) \rightarrow H(s) F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \max(\alpha_f,\alpha_h) < \sigma $$
 
### Conjugación

\begin{equation}
\boxed{{f}^{\ast}(t) \rightarrow F^{\ast}(s^{\ast})}  \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma  
\end{equation}

### Multiplicación

$$ \boxed{\{f g\}(t) \rightarrow \frac{1}{2\pi i} \lim_{\gamma\rightarrow \infty} \int_{\sigma - i\gamma}^{\sigma - i\gamma} F(\tau)G(s-\tau)d\tau}$$

### Integración

\begin{equation}
\boxed{ \int_{-\infty}^{t} f(t) dt \rightarrow \frac{1}{s}F(s) + \frac{1}{s}\int_{-\infty}^{0^{-}} f(t) dt} \;\;\;\;\; \mbox{ROC:}\;\; \max(0,\alpha_f) < \sigma 
\end{equation}


### Teorema del valor inicial

$$ \boxed{f(0^+) = \lim_{s\rightarrow\infty} s F(s)} $$

### Teorema del valor final

$$ \boxed{f(\infty) = \lim_{s\rightarrow 0} s F(s)} $$
si todos los polos de $sF(s)$ están en el semiplano izquierdo.

## Pares de Laplace

$$ \boxed{\mathbf{\delta} \xrightarrow[\text{}]{\mathcal{L}} \mathbf{1}} \;\;\;\;\; \mbox{ROC:}\;\; \forall  \sigma $$

$$ \boxed{ \vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{1}{s}  } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > 0  $$ 

$$ \boxed{ e^{-at}\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{1}{s + a} } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > -a $$

$$ \boxed{ t^{n}\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{n!}{s^{n+1}} } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > 0, n > -1  $$

$$ \boxed{ \cos(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}}  \frac{s}{s^2 + a}  } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > 0 $$

$$ \boxed{ \sin(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{a}{s^2 + a}  } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > 0 $$

$$ \boxed{ e^{-\alpha t}\cos(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{s+\alpha}{(s+\alpha)^2 + a}  } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > -\alpha $$

$$ \boxed{ e^{-\alpha t}\sin(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{a}{(s+\alpha)^2 + a} }\;\;\;\;\; \mbox{ROC:}\;\; \sigma > -\alpha  $$

$$ \boxed{ \cosh(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{s}{s^2 - a^2}  } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > |\alpha| $$

$$ \boxed{ \sinh(at)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} \frac{a}{s^2 - a^2} } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > |\alpha| $$

$$ \boxed{ \ln(t)\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{L}} -\frac{1}{s}[\ln(s) + \gamma] } \;\;\;\;\; \mbox{ROC:}\;\; \sigma > 0 $$

$$ \boxed{ e^{-a|t|} \xrightarrow[\text{}]{\mathcal{L^2}} \frac{2a}{a^2+s^2} } \;\;\;\;\; \mbox{ROC:}\;\; -\alpha < \sigma < \alpha $$

## Propiedades de la Transformada bilateral de Laplace

<div class="alert alert-block alert-info">
Para este resumen consideramos las siguientes transformadas:
$$ f(t) \rightarrow F(s)  \;\;\;\;\; \mbox{ }\;\;\;\;\; g(t) \rightarrow G(s) \;\;\;\;\; \mbox{y}\;\;\;\;\; h(t) \rightarrow H(s) $$

son sus respectivas regiones de converegencia:

$$ \alpha_f < \sigma < \beta_f \;\;\;\;\; \mbox{ }\;\;\;\;\; \alpha_g < \sigma < \beta_g \;\;\;\;\; \mbox{ }\;\;\;\;\; \alpha_h < \sigma < \beta_h$$
</div>

***
### Linealidad
$$ \boxed{af(t)+bg(t) \rightarrow aF(s)+ bG(s)} \;\;\;\;\; \mbox{ROC:}\;\; \max(\alpha_f,\alpha_g) < \sigma < \min(\beta_f,\beta_g)  $$


### Desplazamiento

Para todo $a \in \mathbb{R}$

\begin{equation}
\boxed{f(t-a) \rightarrow e^{-a s} F(s)}  \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f 
\end{equation}

\begin{equation}
\boxed{e^{a t}f(t) \rightarrow  F(s-a)} \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma-a < \beta_f 
\end{equation}


### Escalamiento
Para $a \in \mathbb{R}/\left\{0\right\}$

\begin{equation}
\boxed{f(at) \rightarrow  \frac{1}{|a|}F(\frac{s}{a})} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \frac{\sigma}{a} < \beta_f 
\end{equation}

\begin{equation}
\boxed{f(-t) \rightarrow F(-s)} \;\;\;\;\; \mbox{ROC:}\;\; -\beta_f < \frac{\sigma}{a} < -\alpha_f 
\end{equation}

### Derivada
- **Derivación en frecuencia** 
$$ \boxed{t f(t) \rightarrow -F'(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f $$ 

$$ \boxed{t^{n} f(t) \rightarrow (-1)^nF^{(n)}(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f $$ 

- **Derivación en el tiempo o espacio**
$$ \boxed{f'(t) \rightarrow s F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f  $$

$$ \boxed{f^{(n)}(t) \rightarrow s^n F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f  $$

### Convolución
$$\boxed{(h \ast f)(s) \rightarrow H(s) F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \max(\alpha_f,\alpha_h) < \sigma < \min(\beta_f,\beta_h)  $$
 

### Conjugación

\begin{equation}
\boxed{{f}^{\ast}(t) \rightarrow F^{\ast}(s^{\ast})}  \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \beta_f 
\end{equation}

### Multiplicación

$$ \boxed{\{f g\}(t) \rightarrow \frac{1}{2\pi i} \lim_{\gamma\rightarrow \infty} \int_{\sigma - i\gamma}^{\sigma + i\gamma} F(\tau)G(s-\tau)d\tau} \;\;\;\;\; \;\;\;\;\; \mbox{ROC:}\;\; (\alpha_f+\alpha_g) < \sigma < (\beta_f + \beta_g)   $$

### Integración

\begin{equation}
\boxed{ \int_{-\infty}^{t} f(t) dt \rightarrow \frac{1}{s}F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \max(0,\alpha_f) < \sigma < \beta_f
\end{equation}

\begin{equation}
\boxed{ \int_{t}^{\infty} f(t) dt \rightarrow \frac{1}{s}F(s)} \;\;\;\;\; \mbox{ROC:}\;\; \alpha_f < \sigma < \min(0,\beta_f)
\end{equation}

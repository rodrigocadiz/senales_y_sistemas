## Funciones generalizadas: distribuciones

Cada medición física tiene una incertidumbre finita, ya sea limitada por el aparato experimental o por la naturaleza misma (principio de incertidumbre de Heisenberg). En consecuencia, el valor exacto y exacto de una función $f$ en un valor dado de la coordenada $t$ nunca es realmente accesible experimentalmente. En cambio, si $f$ representa una cantidad física, cada medición de $f$ alrededor de la coordenada $t$ da un promedio ponderado de la forma:

\begin{equation}
\int_{\mathbb{R}}f(t') \varphi(t') dt'   
\label{intro_distr}
\end{equation}

donde $\varphi$ es una "función suave" (el significado de esto se precisará más adelante) que alcanza su punto máximo alrededor de $t'=t$, y que depende de las condiciones experimentales. Por lo tanto, los valores puntuales de $f$ no son relevantes. Por ejemplo, cualquier discontinuidad de $f$ se promedia y, por lo tanto, los valores exactos de $f$ antes y después de un "salto" son irrelevantes. En este contexto, es mejor pensar en $f$ no como una función sino como una "función generalizada" o "distribución" de todos los resultados posibles de la forma anterior para todas las diferentes funciones de ponderación $\varphi$.

Las funciones generalizadas fueron propuestas por Sobolev en 1935. En 1940, Laurent Schwartz extendió estas ideas y formalizó la [teoría de distribuciones](https://es.wikipedia.org/wiki/Teor%C3%ADa_de_distribuciones), por la cual ganó la medalla Fields en 1950.

### ¿Por qué son necesarias las distribuciones?
En muchos problemas físicos, aparecen algunos conceptos que se representan mediante objetos matemáticos idealizados, con características similares a las funciones convencionales, pero que en rigor son cuasi-funciones, en el sentido que no se pueden tratar como funciones matemáticas convencionales. 

Por ejemplo, Dirac introdujo un objeto matemático conocido como impulso, denotado $\delta$, con la siguiente propiedad, conocida como la "propiedad del cedazo":

$$ \int_{-\infty}^{\infty} \delta(x-a) f(x) dx = f(a) $$

El problema es que no existe ningun función matemática convencional que sea la solución de esta ecuación.

La teoría de distribuciones propone el uso de funciones generalizadas o distribuciones, que permiten solucionar este tipo de problemas.

## El espacio de Schwartz

### Funciones bien comportadas (good functions)
En este curso nos interesan funciones que, si bien puedan representar procesos físicos, se comporten adecuadamente. Con eso nos refirimos a funciones que, por ejemplo, sean derivables, que no tengan discontinuidades, y que tengan un soporte compacto, es decir que tengan un valor distinto de cero en un rango acotado de los reales. Una condición menos estricta al rango acotado es que sean [rápidamente decrecientes](https://es.wikipedia.org/wiki/Funci%C3%B3n_de_decrecimiento_r%C3%A1pido), dando origen al [espacio de Schwartz](https://es.wikipedia.org/wiki/Espacio_de_Schwartz).

Una función $f:\mathbb{R} \mapsto \mathbb{C}$ se dice que es rápidamente decreciente si decae hacia cero en el infinito más rápido que cualquier potencia inversa. Más precisamente, $\forall k \in \mathbb{N}$, existe $C_k >0$ tal que

$$ \forall \; t \in \mathbb{R}, \;\;\;\;\;\;  |t|^k |f(t)| \leq C_k $$

o, equivalentemente
$$ \sup_{ t \in \mathbb{R}} |t|^k |f(t)| < \infty $$

tal que $t^k f$ está acotada para todo $ k \in \mathbb{N}$.

### El espacio de Schwartz
El espacio de Schwartz $\mathcal{S}(\mathbb{R};\mathbb{C})$ es el espacio de todas las funciones $\varphi: \mathbb{R} \mapsto \mathbb{C}$ que son infinitamente derivables y que son rápidamente decrecientes. 

$$
\boxed{
\mathcal{S}(\mathbb{R};\mathbb{C}) =\left\{ \varphi \in C^{\infty}(\mathbb{R};\mathbb{C}) \; |  \; \; \forall \alpha, k \in \mathbb{N},\;\;\;  \sup_{t \in \mathbb{R}} | t^{\alpha} \frac{d^k \varphi(t)}{dt^k}   | < \infty    \right\}     }
$$

En otras palabras, el espacio de Schwartz consta de todas las funciones $\varphi$ de modo que $\varphi$ y todas sus derivadas existen sobre $\mathbb{R}$ y decaen a cero en $|t| \rightarrow \infty $ más rápido que cualquier potencia inversa de $t$.

## Distribuciones temperadas

Una distribución es una **función generalizada**, puede considerarse como una "función de funciones". Más precisamente, una distribución es una forma lineal que asocia un número a cada función del espacio de Schwartz. Veremos que la función $\mathbb{1}$, la función escalón de Heavyside, o funciones localmente integrables, como una sinusoide $\sin \omega t$ o $\cos \omega t$, o incluso la $\delta$, pueden asimilarse a ciertas distribuciones. 

### El espacio de distribuciones temperadas $\mathcal{S'}$

El espacio de distribuciones temperadas $\mathcal{S'}$ se define como el espacio de todas las formas lineales y continuas que mapean $\mathcal{S}(\mathbb{R};\mathbb{C})$ a $\mathbb{C}$. En otras palabras, $\mathbf{T}$ es una distribución temperada si el mapa:


$$   \mathbf{T}: \mathcal{S}(\mathbb{R};\mathbb{C})  \mapsto \mathbb{C}    $$

es lineal y 

$$  \varphi_n  \xrightarrow[]{ \mathcal{S} } \varphi \implies      \mathbf{T}(\varphi_n)   \xrightarrow[]{ \mathbb{C} }  \mathbf{T}(\varphi)   $$

Utilizaremos la notación $\langle \mathbf{T}, \varphi \rangle$ en vez de $\mathbf{T}(\varphi)$. 

Esto no debe confundirse con el producto interno en $L^2$, el contexto es claro aquí dado que $\mathbf{T}$ no es una función.

### La distribución delta de Dirac $\delta$

La distribución delta de Dirac $\mathbf{\delta}$ define que tal forma que, para toda función $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

$$
\langle \mathbf{\delta}, \varphi \rangle = \varphi(0) \;\;\; \in \mathbb{C} 
$$

y la distribución delta de Dirac $\mathbf{\delta_a}$ en un punto $a \in \mathbb{R}$ se define como:

$$\langle \mathbf{\delta_a}, \varphi \rangle = \varphi(a) \;\;\; \in \mathbb{C}  \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})
$$

### La distribución escalón de Heaviside

La distribución escalón $\mathbf{u}$ se define de tal manera que, para todo $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

$$
\langle \mathbf{u}, \varphi \rangle = \int_{\mathbb{R}} u(t) \varphi(t) dt = \int_{0}^{\infty} \varphi(t) dt \;\;\; \in \mathbb{C} 
$$

donde $u$ es la función escalón de Heaviside.

### La distribución valor principal

La función $t \mapsto \frac{1}{t}$ no es localmente integrable sobre $\mathbb{R}$. Sin embargo, es posible definir la distribución $\mathbf{\mbox{p.v}\frac{1}{t}}$, llamada valor principal de Cauchy de $1/t$, como la distribución sobre $\mathbb{R}$ definida por la fórmula:

\begin{equation}
\langle \mathbf{\mbox{p.v}\frac{1}{t}}, \varphi \rangle = \lim_{\epsilon \rightarrow 0^+}\left\{  \int_{t<-\epsilon} \frac{\varphi(t)}{t}dt  + \int_{t>\epsilon} \frac{\varphi(t)}{t}dt  \right\} \;\;\; \in \mathbb{C} 
\end{equation}

para toda $\varphi \in \mathcal{D}(\mathbb{R};\mathbb{C})$.


### La distribución seno
La distribución $\mathbf{\mbox{sin}}$ se define, similarmente, para todo $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

$$\langle \mathbf{\mbox{sin}}, \varphi \rangle = \int_{\mathbb{R}} \sin(t) \varphi(t) dt  \;\;\; \in \mathbb{C} 
$$

### La distribución coseno
La distribución $\mathbf{\mbox{cos}}$ se define, similarmente, para todo $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

$$\langle \mathbf{\mbox{cos}}, \varphi \rangle = \int_{\mathbb{R}} \cos(t) \varphi(t) dt  \;\;\; \in \mathbb{C} 
$$


### La distribución uno
La distribución $\mathbf{1}$ se define, similarmente, para todo $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

\begin{equation}
\langle \mathbf{1}, \varphi \rangle = \int_{\mathbb{R}} \varphi(t) dt  \;\;\; \in \mathbb{C} 
\label{one_distribution}
\end{equation}

<a name="distribucion_temperada_generica"></a>
### La distribución de una función temperada genérica

Observamos que, dada una función $h$, es posible definir una distribución temperada $h$ tal que para todo $\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$:

$$\langle \mathbf{\mbox{h}}, \varphi \rangle = \int_{\mathbb{R}} h(t) \varphi(t) dt  \;\;\; \in \mathbb{C} 
$$

dado que la función $\varphi$ es rápidamente decreciente, la función $h$ no necesita ser integrable o cuadrado integrable.

## Transformaciones de distribuciones

### Operaciones lineales y continuas sobre el espacio de Schwartz

Las siguientes son aplicaciones lineales y continuas $L:S(\mathbb{R};\mathbb{C}) \rightarrow S(\mathbb{R};\mathbb{C})$

- Desplazamiento: $\varphi \mapsto  \tau_a \varphi(t) := \varphi (t-a) \;\;\; a \in \mathbb{R}$\\

- Escalamiento: $\varphi \mapsto  \partial_{\lambda} \varphi(t) := \varphi (\lambda t) \;\;\; \lambda \in \mathbb{R}\setminus\left\{ 0 \right\}  $ 

- Diferenciación: $\varphi \mapsto  \frac{d^k \varphi(t)}{d t^k}  \;\;\; k \in \mathbb{N}  $ 

- Multiplicación:   $\varphi \mapsto  X \varphi   $
donde $X$ pertenece al espacio de funciones temperadas $\mathcal{O}_M(\mathbb{R};\mathbb{C})$ 

$$ \mathcal{O}_M(\mathbb{R};\mathbb{C})= \left\{ X \in C^{\infty} (\mathbb{R};\mathbb{C}) , \forall \alpha \in \mathbb{N},  \exists k_{\alpha} \in \mathbb{N} \;\;\; \mbox{tal que} \;\;\; (1+t)^{-k_{\alpha}}\frac{d^{\alpha} X}{d t^{\alpha}} \in L^{\infty}(\mathbb{R};\mathbb{C}) \right\}  $$


### Desplazamiento de una distribución
De lo anterior, vemos que es posible definir el desplazamiento de una distribución en términos del operador transpuesto operando en el espacio de Schwartz. Primero notamos que para todo $h,\varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})$
 
 $$  \int_{\mathbb{R}} (\tau_a h(t)) \varphi(t) dt  =\int_{\mathbb{R}} h(t) (\tau_{-a} \varphi(t) ) dt  \;\;\;\; \forall a \in \mathbb{R} $$
 
Vemos entonces que la distribución $\tau_a \mathbf{h}$ asociada a $\tau_a(h)= h(t-a) \in \mathcal{S}(\mathbb{R};\mathbb{C})$  cumple con:
 
 $$  \langle  \tau_a \mathbf{ h }, \varphi \rangle = \langle \mathbf{h}, \tau_{-a} \varphi  \rangle   $$
 
de tal forma que el operador transpuesto de un desplazamiento en $a$ en $\mathcal{S}(\mathbb{R};\mathbb{C})$ corresponde a un desplazamiento en $-a$ en $\mathcal{S'}$.
Esto se extiende para cualquier distribución temperada $\mathbf{T}$

$$
\boxed{   \langle  \tau_a \mathbf{ (T) }, \varphi \rangle = \langle \mathbf{T}, \tau_{-a} \varphi  \rangle        \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})  }
$$

### Escalamiento de una distribución
También es posible definir el escalamiento mediante el operador transpuesto actuando en el espacio de Schwartz. Notamos que, para todo $h$, $\varphi$ $\in \mathcal{S}(\mathbb{R};\mathbb{C})$
 
  $$  \int_{\mathbb{R}} \left( \partial_a h(t) \right) \varphi(t) dt  =\int_{\mathbb{R}} h(t)\frac{1}{ |a|} \partial_{1/a} \varphi(t) dt  \;\;\;\; \forall a \in \mathbb{R} $$
 
Tal que un escalamiento por $a$ es una distribución temperada $\mathbf{T}$ definida por:

$$
\boxed{   \langle  \partial_a \mathbf{ (T) }, \varphi \rangle = \langle \mathbf{T}, \frac{1}{|a|} \partial_{1/a} \varphi  \rangle \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})  }
$$

### Derivada de una distribución
De la misma manera, es posible definir la derivada de una distribución mediante su operador transpuesto operando en el espacio de Schwartz. De acuerdo a la definición de una distribución genérica, la derivada de una distribución $\mathbf{h}$ debiera ser: 

$$
\langle \mathbf{h'},\varphi \rangle = \int_{-\infty}^{\infty} h'(t)\varphi(t) dt
$$

Notamos que el lado derecho de esta expresión puede ser integrado por partes:

$$
\int_{-\infty}^{\infty} h'(t)\varphi(t) dt = h(t)\varphi(t)|_{-\infty}^{\infty} - \int_{-\infty}^{\infty} h(t)\varphi'(t) dt 
$$

La expresión $h(t)\varphi(t)|_{-\infty}^{\infty}$ es cero, dado que $\varphi$ es una función de Schwartz. Por lo tanto, tenemos que 

$$
\int_{-\infty}^{\infty} h'(t)\varphi(t) dt =  - \int_{-\infty}^{\infty} h(t)\varphi'(t) dt = -\langle \mathbf{h},\varphi' \rangle 
$$

Por lo tanto, concluímos que 
$$
\langle \mathbf{h'},\varphi \rangle = -\langle \mathbf{h},\varphi' \rangle
$$


Podemos extender esta idea para derivadas superiores para todo $h$, $\varphi$ $\in \mathcal{S}(\mathbb{R};\mathbb{C})$
 
  $$  \int_{\mathbb{R}} \left(\frac{d^{\alpha}}{dt^{\alpha}}  h(t) \right)  \varphi(t) dt  =\int_{\mathbb{R}} h(t) (-1)^{\alpha} \frac{d^{\alpha}}{dt^{\alpha}} \varphi(t) dt  \;\;\;\; \forall \alpha \in \mathbb{N} $$
 
Tal que la derivada de orden $\alpha$ es una distribución temperada $\mathbf{T}$ definida por:

$$
\boxed{   \biggr\langle \frac{d^{\alpha}}{dt^{\alpha}} \mathbf{ (T) }, \varphi \biggr\rangle = \biggr\langle \mathbf{T}, (-1)^{\alpha} \frac{d^{\alpha}}{dt^{\alpha}} \varphi  \biggr\rangle        \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})  }
$$

Una distribución temperada puede ser diferenciada en cualquier orden, el resultado sigue siendo una distribución temperada.

### Multiplicación por una función temperada
En general, no es posible definir la multiplicación de dos distribuciones. No es posible incluso definir el producto de una distribución con una función que no es $C^{\infty}$. 
 
Existe, sin embargo, una clase de funciones para las cuales es posible definir la multiplicación por una distribución.

Para cada $X$ $\in \mathcal{O}_M(\mathbb{R};\mathbb{C})$ y cada distribución temperada $\mathbf{T}$, se define una distribución $X \mathbf{T} $ por:

$$
\boxed{   \langle X \mathbf{ T }, \varphi \rangle = \langle \mathbf{T}, X \varphi  \rangle        \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C})  }
$$

En particular, una constante $\alpha \in \mathbb{C}$ pertenece a $\mathcal{O}_M$ tal que una distribución se puede multiplicar por un factor constante. 

#### Ejemplo
Consideremos la multiplicación de $t \in \mathcal{O}_M(\mathbb{R};\mathbb{C})$ y el valor principal $1/t$, $\mathbf{\mbox{p.v}\frac{1}{t}}$:

$$\langle t \mathbf{\mbox{p.v}\frac{1}{t}}, \varphi \rangle =  \langle \mathbf{\mbox{p.v}\frac{1}{t}}, t \varphi \rangle = \mbox{p.v} \int_{\mathbb{R}} t \frac{\varphi(t)}{t} dt =   \int_{\mathbb{R}}  \varphi(t) dt = \langle \mathbf{1}, \varphi \rangle \;\;\;\; \forall \varphi \in \mathcal{S}(\mathbb{R};\mathbb{C}) $$

Finalmente obtenemos un resultado destacable, que la "inversa" de $t$ es el valor principal de $1/t$:

$$ t \mathbf{\mbox{p.v}\frac{1}{t}} = \mathbf {1}   $$

Recordemos que la distribución $1/t$ no existe.

## Cambios de variables en distribuciones

Sea $T_f$ una distribución temperada que proviene de una función integrable $f(x):\mathbb{R}\rightarrow\mathbb{R}$. Nos interesa poder determinar si es posible concebir un "cambio de variables" de $T_f$ hacia una distribución $T_g$ de manera tal que $g(x) = f(w(x))$, donde $w(x):\mathbb{R}\rightarrow\mathbb{R}$. Es decir, nos interesa poder utilizar la distribución derivada de $f(x)$, pero ya no evaluada en $x$, si no en una función de $x$, determinada por $w(x)$.

Dado que $T_f$ proviene de una distribución, podemos escribir la siguiente integral para $T_g$:

$$ \langle T_g, \varphi \rangle = \int_{-\infty}^{\infty} g(x)\varphi(x) dx = \int_{-\infty}^{\infty} f(w(x))\varphi(x) dx  $$

Hacemos el tradicional cambio de variables, $y = w(x)$, lo que implica que $x = w^{-1}(y)$, $dy = w'(x) dx$. Entonces podemos escribir:

 $$ \langle T_g, \varphi \rangle = \int_{-\infty}^{\infty} f(y)\varphi(w^{-1}(y)) \frac{1}{|w'(w^{-1}(y))|} dy  $$
 
 Analizando la integral anterior, vemos que es simplemente
 
$$\boxed{\langle T_f, \frac{\varphi(w^{-1})}{|w'(w^{-1})|} \rangle}  $$

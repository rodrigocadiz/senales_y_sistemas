## Fast Fourier Transform

The discrete Fourier transformation (DFT) can be implemented computationally very efficiently by the [fast Fourier transform (FFT)](https://en.wikipedia.org/wiki/Fast_Fourier_transform). Various algorithms have been developed for the FFT resulting in various levels of computational efficiency for a wide range of DFT lengths. <a name="propiedades"></a>

Al estudiar las propiedades de la Transformada Z, el paralelo con muchas de las propiedades de la Transformada de Fourier son evidentes. Por ejemplo, un desplazamiento de $f$ corresponde a una multiplicación de $\mathcal{Z}\{f\}$ por una exponencial compleja. Para estas propiedades consideraremos funciones derechas y ROC genérica $R$. En general, consideraremos las propiedades de la Transformada Z bilateral, a menos que se indique lo contrario.

### Linealidad

La Transformada de Z es una aplicación lineal, por lo que cumple con homogeneidad, para todo $a \in \mathbb{R}$

$$ \boxed{a f[n] \rightarrow a F(z)} $$ 

y superposición

$$ \boxed{f[n]+g[n] \rightarrow F(z)+G(z)} $$

cuya ROC contiene al menos la intersección de $R_f$ y $R_g$.

### Desplazamiento
Para todo $m \in \mathbb{R}^+$:

\begin{equation}
\boxed{f[n-m] \rightarrow z^{-m} F(z)} 
\end{equation}

con ROC = $R$, con la posible excepción de $z=0$ o $z=\infty$.

Para el caso de la Transformada Z unilateral, esta propiedad es:

\begin{equation}
\boxed{f[n-1] \rightarrow z^{-1} F(z) + f[-1]} 
\end{equation}

y

\begin{equation}
\boxed{f[n+1] \rightarrow zF(z) - zf[0]} 
\end{equation}



### Escalamiento
Para $a \in \mathbb{C}$: 


\begin{equation}
\boxed{a^n f[n] \rightarrow F\left(\frac{z}{a}\right)}
\end{equation}

con ROC = $|a|R$.

### Reversión o traslape

\begin{equation}
\boxed{f[-n] \rightarrow F\left(\frac{1}{z}\right)}
\end{equation}

con ROC = $R^{-1}$.

### Estiramiento
Si $r \in \mathbb{Z}$:

\begin{equation}
\boxed{f_k[n] = \begin{cases} f[r], & n = rk \\ 0, & n \neq kr \end{cases} \rightarrow F\left(z^{k}\right)}
\end{equation}

con ROC = $R^{1/k}$, es decir, el conjunto de los puntos $z^{1/k}$, donde $z \in R$.


### Derivada 

\begin{equation}
\boxed{n f[n] \rightarrow -z \frac{dF(z)}{dz}}
\end{equation}

con ROC = $R$.


### Acumulación o integración 

\begin{equation}
\boxed{\sum_{k=-\infty}^{n} f[k] \rightarrow  \frac{1}{1-z^{-1}}F(z)}
\end{equation}

con ROC al menos la intersección entre $R$ y $|z| > 1$.

### Primera diferencia

\begin{equation}
\boxed{f[n] - f[n-1] \rightarrow  (1-z^{-1})F(z)}
\end{equation}

con ROC al menos la intersección entre $R$ y $|z| > 0$.

Para el caso de la Transformada Z unilateral, tenemos:

\begin{equation}
\boxed{f[n] - f[n-1] \rightarrow  (1-z^{-1})F(z) - f[-1]}
\end{equation}

### Convolución

\begin{equation}
\boxed{\{f \ast g \}[n] \rightarrow  F(z)G(z) }
\end{equation}

donde la ROC es al menos la intersección de $R_f$ y $R_g$.

### Conjugación

\begin{equation}
\boxed{f^{\ast}[n] \rightarrow  F^{\ast}(z^{\ast}) }
\end{equation}

con ROC = $R$.

## Comportamiento asintótico
Tal como es el caso de la Transformada de Laplace función discreta $f[n]$ en el infinito (o en el origen $n=0$) y el de su Transformada Z en $|z|=0$ (o $|z|=+\infty)$. 

### Teorema del valor final

$$\lim_{n\rightarrow \infty} f[n] = \lim_{z \rightarrow 1}(z-1)F(z) $$

Lo anterior requiere que los polos de $(z-1)F(z)$ estén al interior del círculo unitario.


### Teorema del valor inicial 
Para una secuencia $f[n]$ causal:

$$ f[0] = \lim_{z \rightarrow +\infty} F(z) $$

## Pares comunes de la Transformada Z


| $f[n]$ | $F(z) = \mathcal{Z} \{ f[n] \}$ | Región de convergencia (ROC) |
|:-----|:-------|:-----|
| $\delta[n]$ | $1$ | Todo el plano-$z$ |
| $u[n]$ | $\frac{1}{1-z^{-1}}$ | $|z| > 1$ |
| $n u[n]$ | $\frac{z^{-1}}{(1-z^{-1})^2}$ | $|z| > 1$ |
| $a^{n} u[n]$ | $\frac{1}{1 - az^{-1}}$ | $|z| > |a|$ |
| $-a^{n} u[-n-1]$ | $\frac{1}{1 - az^{-1}}$ | $|z| < |a|$ |
| $\sin(\omega_0 n) u[n]$ | $\frac{\sin(\omega_0)z^{-1}}{1 - 2 \cos(\omega_0)z^{-1} + z^{-2}}$ | $|z| > 1$ |
| $\cos(\omega_0 n) u[n]$ | $\frac{1 - \cos(\omega_0)z^{-1}}{1 - 2 \cos(\omega_0)z^{-1} + z^{-2}}$ | $|z| > 1$ |
| $r^n\sin(\omega_0 n) u[n]$ | $\frac{\sin(\omega_0)z^{-1}}{1 - 2r \cos(\omega_0)z^{-1} + r^2z^{-2}}$ | $|z| > r$ |
| $r^n\cos(\omega_0 n) u[n]$ | $\frac{1 - r\cos(\omega_0)z^{-1}}{1 - 2 r\cos(\omega_0)z^{-1} + r^2z^{-2}}$ | $|z| > r$ |

donde $a \in \mathbb{C}$, $\omega_0,r \in \mathbb{R}$, $n \in \mathbb{N}$ y $u[n]$ es el escalón unitario.

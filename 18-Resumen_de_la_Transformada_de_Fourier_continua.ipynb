## Resumen de Propiedades de la Transformada de Fourier

$$ f(t) \rightarrow F(u)  \;\;\;\;\; \mbox{ }\;\;\;\;\; g(t) \rightarrow G(u) \;\;\;\;\; \mbox{y}\;\;\;\;\; h(t) \rightarrow H(u) $$

### Linealidad

$$ \boxed{a f(t) \rightarrow a F(u)} $$ 

$$ \boxed{f(t)+g(t) \rightarrow F(u)+G(u)} $$

### Dualidad
$$ \boxed{F(t) \rightarrow f(-u)} $$


\begin{equation}
\boxed{\mathcal{F}\{ \mathcal{F}\{f\} \}(u) = f(-u)}
\end{equation}

<a name="desplazamiento"></a>
### Desplazamiento

Para todo $a \in \mathbb{R}$

\begin{equation}
\boxed{f(t-a) \rightarrow e^{-i 2 \pi a u} F(u)}  \;\;\;\;\; \mbox{y}\;\;\;\;\;  \boxed{e^{i 2 \pi at}f(t) \rightarrow  F(u-a)}
\end{equation}

<a name="escalamiento"></a>
### Escalamiento
Para $a \in \mathbb{R}/\left\{0\right\}$

\begin{equation}
\boxed{f(at) \rightarrow  \frac{1}{|a|}F(\frac{u}{a})}
\end{equation}


### Derivada
- **Derivación en frecuencia**. 
$$ \boxed{t f(t) \rightarrow \frac{i}{2 \pi} F'(u)} $$ 

$$\mathcal{F}\left\{ (-i 2\pi t)^{\alpha} \mathbf{T} \right\} = D^{\alpha}\mathcal{F}\left\{ \mathbf{T} \right\}  \;\;\; \alpha \in \mathbb{N}$$

(no válida en $L^2$).

- **Derivación en el tiempo o espacio**. 
$$ \boxed{f'(t) \rightarrow i 2 \pi u F(u)} $$


$$\mathcal{F}\left\{ D^{\alpha} \mathbf{T} \right\} = (i 2 \pi u)^{\alpha} \mathcal{F}\left\{ \mathbf{T} \right\}  \;\;\; \alpha \in \mathbb{N}$$

### Convolución
 $$ G(u) = \mathcal{F}\{ h \ast f\} (u) = H(u) F(u) $$
 

### Multiplicación

$$ \boxed{\{f g\}(t) \rightarrow \{F \ast G\}(u)}$$

### Fórmula de Transferencia

\begin{equation}
\boxed{ \int_{\mathbb{R}}   \mathcal{F}\{f\} (t) g(t) dt = \int_{\mathbb{R}}   f(t) \mathcal{F}\{g\}(t) dt}  
\end{equation}

### Plancherel-Parseval

Si $f,g \in   L^1(\mathbb{R};\mathbb{C}) \cap L^2(\mathbb{R};\mathbb{C})$, entonces sus Transformadas de Fourier están en $L^2(\mathbb{R};\mathbb{C}) $ y se tiene que:

\begin{equation}
\boxed{(\mathcal{F}\{f\}, \mathcal{F}\{g\} ) = (f,g ); \; \;\;\;\;\; \| \mathcal{F}\{f\} \|_{L^2}^2  = \|  f \|_{L^2}^2   }  
\end{equation}

$$ (f, f) = \int_{\mathbb{R}} |f(t)|^2 dt = \int_{\mathbb{R}} |F(u)|^2 du $$


### Desigualdad de Heisenberg
Sea $f \in   L^2(\mathbb{R};\mathbb{C}) $ y  $f \in   C^{1}(\mathbb{R};\mathbb{C}) $ tal que $tf, df/dt \in  L^2(\mathbb{R};\mathbb{C})$.

\begin{equation}
\left( \int_{\mathbb{R}} t^2 |f(t)|^2 dt \right)^{1/2} \left( \int_{\mathbb{R}} u^2 | \mathcal{F}\{f \}(u) |^2 du   \right)^{1/2} \geq  \left\| f \right\|^2_{L^2}
\end{equation}

o, equivalentemente:
$$  \boxed{\left\| t f \right\|_{L^2}  \left\| u \mathcal{F}\{f\} \right\|_{L^2} \geq \left\|  f \right\|_{L^2}^2} $$

## Pares de Fourier


$$ \boxed{\mathbf{\mbox{sinc(t)}} \longleftrightarrow \sqcap(u)} $$

$$ \boxed{\mathbf{\wedge}(t) \longleftrightarrow \mathbf{\mbox{sinc}^2}\left( u \right)} $$

$$ \boxed{ e^{-at}\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{F}} \left( \frac{1}{a+ i 2\pi u} \right) } $$

$$ \boxed{ e^{-a|t|} \xrightarrow[\text{}]{\mathcal{F}} \frac{2a}{a^2+(2\pi u)^2} } $$

$$ \boxed{\frac{1}{a+i 2 \pi t} \rightarrow e^{a u}\vcenter{{\large \ulcorner}}(-u)}   $$

$$ \boxed{ \frac{2a}{a^2+t^2} \xrightarrow[\text{}]{\mathcal{F}} 2\pi e^{-a 2 \pi |u|}  } $$

$$ \boxed{\mathbf{\delta}  \longleftrightarrow \mathbf{1}}$$
 
$$ \boxed{\mathbf{e^{i 2 \pi a t}}  \longleftrightarrow \delta_a} $$
 
$$ \boxed{\mathbf{\mbox{cos}(2 \pi a t)} \rightarrow \frac{1}{2} (\delta_a + \delta_{-a})} $$
 
$$ \boxed{\mathbf{\mbox{sin}(2 \pi a t)} \rightarrow \frac{1}{2i} (\delta_a - \delta_{-a})} $$

$$ \boxed{\mathbf{\vcenter{{\large \ulcorner}}} \rightarrow \frac{1}{2}\left(\delta + \frac{1}{i \pi u}\right)} $$ 

$$ \boxed{\mathbf{\mbox{sgn}} \rightarrow \frac{\mathbf{1}}{i \pi u}} $$

$$ \boxed{\mathbf{{\bot \!\! \bot \!\! \bot}_a} \rightarrow \frac{1}{a} \mathbf{{\bot \!\! \bot \!\! \bot}_{1/a}}} $$

## Transformada de Fourier en $L^1$

Para cada función integrable $f \in L^1(\mathbb{R};\mathbb{C})$, la Transformada de Fourier $\mathcal{F}$ de $f$ es la función $\mathcal{F}\{f\}$ que se define sobre $\mathbb{R}$ mediante la integral:

\begin{equation}
\boxed{ F(u) = \mathcal{F}\{f\} (u) = \int_{\mathbb{R}} f(t) e^{-i 2 \pi u t} dt }
\end{equation}

La integral existe si y solo si $f \in L^1(\mathbb{R};\mathbb{C})$.

Una notación que usaremos frecuentemente es la siguiente:

$$ \boxed{f(t) \xrightarrow[\text{}]{\mathcal{F}} F(u)} $$

o simplemente

$$ \boxed{f(t) \rightarrow F(u)} $$

que debe entenderse como que la Transformada de Fourier de $f(t)$ es $F(u)$.

Si bien $f$ no necesita ser definida para todo valor de $t \in \mathbb{R}$ (consideramos $f$ y todas sus funciones equivalente como el mismo elemento de $L^1$), eso no es el caso para la Transformada de Fourier, la cual se define únivocamente para todo valor de $u \in \mathbb{R}$ mediante el valor de la integral.


### Teorema de Riemman-Lebesgue
<div class="alert alert-block alert-info">
Para toda $f \in L^1(\mathbb{R};\mathbb{C})$, su Transformada de Fourier $\mathcal{F}\{f\}$ es una función continua y acotada ($\mathcal{F}\{f\} \in C_b(\mathbb{R};\mathbb{C}) $). De hecho, $\mathcal{F}\{f\}$ está acotada por la norma $L_1$ de $f$, $ | \mathcal{F}\{f\} (u) | \leq \| f\|_{L^1} $ para todo $u \in \mathbb{R}$. Más aún, $\mathcal{F}\{f\}$ decae a cero cuando $|u| \rightarrow \infty$. Esto se conoce como el <b>Teorema de Riemann-Lebesgue</b>.

La Transformada de Fourier $\mathcal{F}: f \mapsto \mathcal{F}\{f\}$ es un mapa lineal desde $L^1(\mathbb{R};\mathbb{C})$ hacia $C_0(\mathbb{R};\mathbb{C})$, el espacio de funciones continuas que tienden a cero en el infinito.  Dado que un elemento de $C_0(\mathbb{R};\mathbb{C})$ no necesariamente pertenece a $L^1(\mathbb{R};\mathbb{C})$, no es posible, en general, definir la Transformada de Fourier de $\mathcal{F}\{f\}$. Esto es una importante limitación de la Transformada de Fourier cuando está restringida solo a funciones integrables. 
</div>

***
### La Transformada de Fourier es una aplicación continua

$$ \mathcal{F}: L^1(\mathbb{R};\mathbb{C}) \mapsto C_0 (\mathbb{R};\mathbb{C})$$ 

Dado que cada función $\phi$ que decae a cero en el infinito ($\phi$ $\in$ $C_0(\mathbb{R};\mathbb{C})$) es acotada, podemos definir su norma $C_0$ mediante la norma $L_{\infty}$:

$$ \|\phi\|_{C_0} =\|\phi\|_{\infty}= \sup_{t \in \mathbb{R}}  |\phi(t) | $$

Lo que implica que la Transformada de Fourier es una aplicación continua  $\mathcal{F}: L^1(\mathbb{R};\mathbb{C}) \mapsto C_0 (\mathbb{R};\mathbb{C})$. Esto significa que si $f_n \xrightarrow[\text{}]{L^{1}} f$, entonces $ \mathcal{F}\{f_n\} \xrightarrow[]{L^\infty} \mathcal{F}\{f\} \in C_0(\mathbb{R},\mathbb{C})$. 

En efecto, tenemos

$$ | \mathcal{F}\{f_n\}(u)-\mathcal{F}\{f\}(u)|= | \mathcal{F}\{f_n-f\}(u)|    \leq  \| f_n-f \|_{L^1} \;\;\; \forall u \in \mathbb{R} $$

tal que 
$$\sup_{u \in \mathbb{R}} |\mathcal{F}\{f_n\}(u)-\mathcal{F}\{f\}(u)|  \leq \| f_n-f\|_{L^1}\;\;\;\;\;\; \forall n \in \mathbb{N} $$ 

y dado que  $\lim_{n\rightarrow \infty}\| f_n-f \|_{L^1} = 0   $

$$\lim_{n\rightarrow \infty} \sup_{u \in \mathbb{R}} |\mathcal{F}\{f_n\}(u)-\mathcal{F}\{f\}(u)| = \lim_{n\rightarrow \infty} \|\mathcal{F}\{f_n\}-\mathcal{F}\{f\}\|_{\infty} = 0$$ 

lo que prueba que $\mathcal{F}\{f_n\} $ converge uniformemente a $\mathcal{F}\{f\}$ en $C_0 (\mathbb{R};\mathbb{C})$.


## Otras definiciones

Existen otras formas de definir la Transformada de Fourier. Una de las más utilizadas es:

\begin{equation}
\boxed{ \hat f (\omega) = \frac{1}{\sqrt{2\pi}} \int_{\mathbb{R}} f(t) e^{-i\omega t} dt }
\end{equation}

o bien

\begin{equation}
\boxed{ \hat f (\omega) =  \int_{\mathbb{R}} f(t) e^{-i\omega t} dt }
\end{equation}

En esta definición, se utiliza la variable $\omega = 2 \pi u$ y un factor de amplificación que se puede escoger arbitrariamente. Se suele escoger $1$, $1/\sqrt{2\pi}$ o $1/2\pi$.

### Ejemplo: Transformada de Fourier de un rect

Consideremos la función rectangular $\sqcap$ de ancho 1. Claramente, $ \sqcap \in  L^1(\mathbb{R};\mathbb{C})$, porque
 

$$\|\sqcap  \|_{L^1} = \int_{-1/2}^{1/2}  dt = 1 < \infty$$
 
Esto implica que esta función tiene Transformada de Fourier. El valor preciso en la discontinuidad $t= \pm \frac{1}{2}$ no es relevante, podemos escoger $ \sqcap(\pm 1/2) = 1$ (o cualquier otro valor) y la función resultante sería equivalente a $ \sqcap$ en $ L^1(\mathbb{R};\mathbb{C})$ y obtendríamos la misma Transformada de Fourier.  Calculemos la Transformada de Fourier, para $u \neq 0$:

$$ \mathcal{F}\{\sqcap\}(u) = \int_{-1/2}^{1/2} e^{-i 2 \pi u t} dt = 2 \int_{0}^{1/2} \cos(2 \pi u t)dt $$


$$ \mathcal{F}\{\sqcap\} (u)  =  2 \frac{\sin(2 \pi u/2)}{2 \pi u}  =  \frac{\sin(\pi u)}{\pi u} = \mbox{sinc}(u)$$

Además

$$ \mathcal{F}\{\sqcap\} (0) = \int_{-1/2}^{1/2} dt = 1  $$

Con este cálculo podemos establecer nuestro primer par de Fourier:

$$ \boxed{\sqcap(t) \xrightarrow[\text{}]{\mathcal{F}} \mbox{sinc}(u)} $$

Aún cuando $ \sqcap(t)$ tiene discontinuidades, su Transformada de Fourier es continua y acotada para todo $u \in \mathbb{R}$  (como lo indica el teorema de  Riemann-Lebesgue), en efecto:

$$   \sup_{u \in \mathbb{R}} |   \mbox{sinc}(u)  | = 1  \leq \| \sqcap  \|_{L^1} = 1 $$

La Transformada de Fourier de $ \sqcap$ tiende a cero cuando $|u| \rightarrow \infty$. Sin embargo, no es una función integrable, dado que la integral sobre $\mathbb{R}$ de $|\sin \pi t |/\pi t$ diverge. <font color="red">No es posible definir la Transformada de Fourier de un sinc, tal como la hemos concebido hasta ahora</font>.

Para un rect de ancho $\Delta t = a$, mientras más ancho sea, más concentrada estará su Transformada de Fourier en torno a $u = 0$, dado que su ancho $\Delta u$ es inversamente proporcional a $a$. Este es un resultado general que se conoce como el **principio de incertidumbre**.

Veamos como calcular esta transformada mediante el paquete de matemática simbólica sympy:

```python
import sympy as sym
import matplotlib.pyplot as plt

t,u = sym.symbols("t u")

a = sym.Symbol('a', positive=True)
f = sym.Heaviside(t+1/2) - sym.Heaviside(t-1/2)
F = sym.fourier_transform(f, t, u)
F
```

Grafiquemos este par de Fourier:

```python
def move_sympyplot_to_axes(p, ax):
    backend = p.backend(p)
    backend.ax = ax
    backend.process_series()
    backend.ax.spines['right'].set_color('none')
    backend.ax.spines['bottom'].set_position('zero')
    backend.ax.spines['top'].set_color('none')
    plt.close(backend.fig)
    

p1 = sym.plot(f,(t, -2,2),show=False,xlabel='$t$',ylabel='',title='$f(t)$')
p2 = sym.plot(sym.re(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',title='$F(u)$',legend=True)
p3 = sym.plot(sym.im(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',line_color='r')
p2.extend(p3)

p2[0].label = '$\operatorname{Re}\{F(u)\}$'
p3[0].label = '$\operatorname{Im}\{F(u)\}$'


fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(15, 5))
move_sympyplot_to_axes(p1, ax)
move_sympyplot_to_axes(p2, ax2)
plt.show()
```

### Ejemplo: Transformada de Fourier de una exponencial lateral

Consideremos la función lateral

$$ f(t) = e^{-at}\vcenter{{\large \ulcorner}}(t)$$ 

donde $u(t)$ es el escalón y $a>0$. Se llama lateral porque solo decrece hacia un lado del eje real. Podemos verificar que $f \in L^1(\mathbb{R};\mathbb{C}) $ dado que

$$  \| f  \|_{L^1}  =  \int_{0}^{\infty} | e^{-at} | dt = \frac{1}{a}  $$

Su Transformada de Fourier es

$$ \mathcal{F}\{f\}(u) =  \int_{0}^{\infty}  e^{-at} e^{-i 2\pi u t} dt = \left( \frac{1}{a+ i 2\pi u} \right)  $$

Vemos que $\mathcal{F}\{f\}$ es continua y acotada. En efecto,

$$ | \mathcal{F}\{f\}(u) | = \frac{1}{\sqrt{a^2 + (2 \pi u) ^2}}  \leq \frac{1}{a} = \| f  \|_{L^1}   \;\;\;\; \forall u \in \mathbb{R}$$

y tiende a cero en el infinito:

$$ \lim_{| u | \rightarrow \infty }  |\mathcal{F}\{f\} (u) |  = 0 $$

en acordancia con el teorema de Riemann-Lebesgue.

Nuestro segundo par de Fourier es :

$$ \boxed{ e^{-at}\vcenter{{\large \ulcorner}}(t) \xrightarrow[\text{}]{\mathcal{F}} \left( \frac{1}{a+ i 2\pi u} \right) } $$


Nuevamente tenemos que $\mathcal{F}\{f\}$ no es integrable, porque la integral sobre $\mathbb{R}$ de $1/\sqrt{(2\pi u)^2 + a^2}$ diverge, por lo que $\mathcal{F}\{f\}$ no tiene Transformada de Fourier como la hemos definido.

Veamos como calcular esta transformada mediante el paquete de matemática simbólica sympy:

```python
import sympy as sym
import matplotlib.pyplot as plt

t,u = sym.symbols("t u")

a = sym.Symbol('a', positive=True)
f = sym.exp(-a*t)*sym.Heaviside(t)
F = sym.fourier_transform(f, t, u)
F
```

Grafiquemos este par de Fourier para el caso $a=2$:

```python
def move_sympyplot_to_axes(p, ax):
    backend = p.backend(p)
    backend.ax = ax
    backend.process_series()
    backend.ax.spines['right'].set_color('none')
    backend.ax.spines['bottom'].set_position('zero')
    backend.ax.spines['top'].set_color('none')
    plt.close(backend.fig)
    
    
f = f.subs(a,2)
F = F.subs(a,2)

p1 = sym.plot(f,(t, -2,2),show=False,xlabel='$t$',ylabel='',title='$f(t)$')
p2 = sym.plot(sym.re(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',title='$F(u)$',legend=True)
p3 = sym.plot(sym.im(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',line_color='r')
p2.extend(p3)

p2[0].label = '$\operatorname{Re}\{F(u)\}$'
p3[0].label = '$\operatorname{Im}\{F(u)\}$'


fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(15, 5))
move_sympyplot_to_axes(p1, ax)
move_sympyplot_to_axes(p2, ax2)
plt.show()
```

### Ejemplo: Transformada de Fourier de una exponencial  doble

Consideremos ahora 

$$f(t)= e^{- a|t|}$$

la cual $\in L^1(\mathbb{R};\mathbb{C}) $ para $a>0$, conocida como una exponencial doble.

Su Transformada de Fourier es:

$$ \mathcal{F}\{f\}(u) = \int_{\mathbb{R}}  e^{-i 2\pi u t}e^{-a|t|} dt =  \left\{\int_{0}^{\infty}  e^{-i 2\pi u t}e^{-at} dt  +   \int_{-\infty}^{0}  e^{-i 2\pi u t}e^{at} dt \right\}   $$

  
$$ \mathcal{F}\{f\}(u) = \left\{  -\frac{1}{a+i 2\pi u}e^{-(a+i 2\pi u)t}\biggr\rvert_{0} ^{\infty}  +   \frac{1}{a-i 2\pi u}e^{(a-i 2\pi u)t}\biggr\rvert_{-\infty}^{0}  \right\}   = \left\{  \frac{1}{a+i 2\pi u} +   \frac{1}{a-i 2\pi u} \right\}  $$


Vemos que la Transformada de Fourier de una exponencial doble:


$$  \mathcal{F}\{f\}(u) = \frac{2a}{a^2+(2\pi u)^2}  \;\; \in L^1(\mathbb{R};\mathbb{C})  $$

Esta función también se conoce la distribución de probabilidades de Cauchy-Lorentz.

Por lo tanto, nuestro tercer par de Fourier es :

$$ \boxed{ e^{-a|t|} \xrightarrow[\text{}]{\mathcal{F}} \frac{2a}{a^2+(2\pi u)^2} } $$

Observamos esta vez que $f(t)=e^{- a|t|} $ decae como $1/u^2$ hacia el infinito y es integrable, lo que no era el caso de la exponencial lateral, cuya discontinuidad abrupta en $t=0$ se manifesta como un Transformada de Fourier que decae lento, como $1/u$ hacia el infinito. 

Intuimos que mientras mas suave sea la función $f$, más rápido decaerá su Transformada de Fourier en el infinito.

Definiendo el ancho $\Delta t$ de $f$ como el ancho completo en la mitad del máximo, se obtiene $\Delta t= 2 \ln 2 / a$. Similarmente, el ancho de $\mathcal{F}\{f\}$ es $\Delta \pi u = a$. Confirmamos nuevamente el resultado general, mientras más grande el ancho $\sim 1/a$ de la exponencial sobre, más angosta será su Transformada de Fourier. 

Veamos como calcular esta transformada mediante el paquete de matemática simbólica sympy:

```python
import sympy as sym
import matplotlib.pyplot as plt

t,u = sym.symbols("t u")

a = sym.Symbol('a', positive=True)
f = sym.exp(-a*abs(t))
F = sym.fourier_transform(f, t, u)
F
```

Grafiquemos este par de Fourier para el caso $a=2$:

```python
def move_sympyplot_to_axes(p, ax):
    backend = p.backend(p)
    backend.ax = ax
    backend.process_series()
    backend.ax.spines['right'].set_color('none')
    backend.ax.spines['bottom'].set_position('zero')
    backend.ax.spines['top'].set_color('none')
    plt.close(backend.fig)
    
    
f = f.subs(a,2)
F = F.subs(a,2)

p1 = sym.plot(f,(t, -2,2),show=False,xlabel='$t$',ylabel='',title='$f(t)$')
p2 = sym.plot(sym.re(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',title='$F(u)$',legend=True)
p3 = sym.plot(sym.im(F),(u, -5,5),show=False,xlabel='$u$',ylabel='',line_color='r')
p2.extend(p3)

p2[0].label = '$\operatorname{Re}\{F(u)\}$'
p3[0].label = '$\operatorname{Im}\{F(u)\}$'


fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(15, 5))
move_sympyplot_to_axes(p1, ax)
move_sympyplot_to_axes(p2, ax2)
plt.show()
```

### Ejemplo: Transformada de Fourier de la distribución de Lorentz-Cauchy

Dado que en el caso anterior la Transformada de Fourier de la exponencial doble es integrable, podemos calcular su Transformada de Fourier:

Consideremos la función
$$f(t) = \frac{2a}{a^2+t^2} \in L^1(\mathbb{R};\mathbb{C}) \mbox{ con } a >0 $$

Su Transformada de Fourier es:

$$ \mathcal{F}\{f\}(u) = \int_{\mathbb{R}} \frac{2a}{a^2+t^2}e^{-i 2 \pi u t} dt $$

Esta integral no es trivial de encontrar, pero puede demostrarse que: 

$$ \mathcal{F}\{f\}(u) = 2\pi e^{-a 2 \pi |u|} $$

Nuestro cuarto par es:

$$ \boxed{ \frac{2a}{a^2+t^2} \xrightarrow[\text{}]{\mathcal{F}} 2\pi e^{-a 2 \pi |u|}  } $$

Observamos un resultado importante con este par de funciones, la exponencial doble y la distribución de Lorentz-Cauchy. <font color="red">Para estas dos funciones, al aplicar la Transformada de Fourier dos veces da como resultado la función original!</font>

Veremos más adelante que esta es una propiedad general. Para cualquier función integrable $f$ tal que $\mathcal{F}\{f\}$ es integrable, al aplicar la Transformada de Fourier dos veces otorga $f(-t)$  (en otras palabras, la Transformada de Fourier $\mathcal{F}$ es "casi" su propia inversa). Dado que la exponencial doble es un función par, la variable independiente no se invierte en este caso.  

Veamos como calcular esta transformada mediante el paquete de matemática simbólica sympy:

```python
import sympy as sym
import matplotlib.pyplot as plt
from mpmath import pi

t,u = sym.symbols("t u")

a = sym.Symbol('a', positive=True)
f = (2*a)/(a**2+t**2)
F = sym.fourier_transform(f, t, u)
F
```

Grafiquemos este par de Fourier para el caso $a=2$:

```python
def move_sympyplot_to_axes(p, ax):
    backend = p.backend(p)
    backend.ax = ax
    backend.process_series()
    backend.ax.spines['right'].set_color('none')
    backend.ax.spines['bottom'].set_position('zero')
    backend.ax.spines['top'].set_color('none')
    plt.close(backend.fig)
    
    
f = f.subs(a,2)
F = 2*pi*sym.exp(-4*pi*abs(u))

p1 = sym.plot(f,(t, -20,20),show=False,xlabel='$t$',ylabel='',title='$f(t)$')
p2 = sym.plot(sym.re(F),(u, -10,10),show=False,xlabel='$u$',ylabel='',title='$F(u)$',legend=True)
p3 = sym.plot(sym.im(F),(u, -10,10),show=False,xlabel='$u$',ylabel='',line_color='r')
p2.extend(p3)

p2[0].label = '$\operatorname{Re}\{F(u)\}$'
p3[0].label = '$\operatorname{Im}\{F(u)\}$'


fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(15, 5))
move_sympyplot_to_axes(p1, ax)
move_sympyplot_to_axes(p2, ax2)
plt.show()
```

## Transformada Inversa

Vimos anteriormente que la distribución de Lorentz-Cauchy y la exponencial doble son transformadas de Fourier duales. Lo mismo pasa con una función Gaussiana, como veremos a continuación. Es tentador pensar que esta es una propiedad general de la transformada de Fourier, sin embargo, hemos visto que la transformada de Fourier $ F(u) $ de una función integrable $ f(t) $ no necesariamente es integrable. Por lo tanto, no es posible en general aplicar la operación dual asociada a la Transformada de Fourier. Sin embargo, para una clase reducida de funciones integrables, es posible mostrar que la Transformada de Fourier de $\mathcal{F}$ está estrictamente relacionada con $f$.

Sea $f \in  L^1(\mathbb{R};\mathbb{C})$ tal que su Transformada de Fourier $\mathcal{F}\{f\}$ es integrable ($\mathcal{F}\{f\} \in  L^1(\mathbb{R};\mathbb{C})$). Entonces, para casi todo $t \in \mathbb{R}$:

\begin{equation}
\boxed{ \int_{\mathbb{R}} e^{+i 2 \pi u t} \mathcal{F}\{f\}(u) du = f (t)}
\end{equation}

o bien 

\begin{equation}
\int_{\mathbb{R}} e^{+i 2 \pi u t} F(u) du = f (t)
\end{equation}

o

\begin{equation}
\mathcal{F^{-1}}\{F(u)\} = f (t)
\end{equation}

Si reemplazamos $t$ por $-t$ en la primera ecuación, tenemos que:

\begin{equation}
\int_{\mathbb{R}} e^{-i 2 \pi u t} \mathcal{F}\{f\}(u) du = f (-t)
\end{equation}

Reconocemos en el lado izquierdo la Transformada de Fourier de $\mathcal{F}\{ f\}$, en donde los roles de $t$ y $u$ se han intercambiado. Esto se puede generalizar como:

\begin{equation}
\boxed{\mathcal{F}\{ \mathcal{F}\{f\} \}(u) = f(-u)}
\end{equation}

donde $u$ es la variable independiente de $f$. Este resultado también se conoce como la fórmula inversa o la propiedad de la dualidad.

Otras formas de escribir esta propiedad pueden ser:

$$ F(t) \rightarrow f(-u) $$
$$ \mathcal{F}\{f\}(-u) = \mathcal{F^{-1}}\{f\}(u) $$
$$ \mathcal{F^{-1}}\{f\}(-t) = \mathcal{F}\{f(t)\} $$

## Propiedades de la Transformada de Fourier en $L^1$

Para efecto de las siguientes propiedades, sean $f, g, h \in L^1(\mathbb{R};\mathbb{C}) $ y $F(u),G(u),H(u)$ sus Transformadas de Fourier, es decir consideramos los pares de Fourier:

$$ f(t) \rightarrow F(u),  \;\;\;\;\; \mbox{ }\;\;\;\;\; g(t) \rightarrow G(u) \;\;\;\;\; \mbox{y}\;\;\;\;\; h(t) \rightarrow H(u) $$

### Linealidad

La Transformada de Fourier es una aplicación lineal, por lo que cumple con homogeneidad, para todo $a \in \mathbb{R}$

$$ \boxed{a f(t) \rightarrow a F(u)} $$ 

y superposición

$$ \boxed{f(t)+g(t) \rightarrow F(u)+G(u)} $$

### Dualidad
Bajo ciertas condiciones, se tiene que

$$ \boxed{F(t) \rightarrow f(-u)} $$

tal como se discutió anteriormente al definir la transformada inversa.

### Desplazamiento

Para todo $a \in \mathbb{R}$

\begin{equation}
\boxed{f(t-a) \rightarrow e^{-i 2 \pi a u} F(u)}  \;\;\;\;\; \mbox{y}\;\;\;\;\;  \boxed{e^{i 2 \pi at}f(t) \rightarrow  F(u-a)}
\end{equation}

o, en términos del operador de desplazamiento $\tau$


\begin{equation}
\tau_a f(t) \rightarrow e^{-i 2 \pi a u} F(u)  \;\;\;\;\; \mbox{y}\;\;\;\;\;  e^{i 2 \pi at}f(t) \rightarrow  \tau_a F(u)  
\end{equation}


### Escalamiento
Para $a \in \mathbb{R}/\left\{0\right\}$ definimos $ \partial_a f(t) = f_a(t) = f(at) $. Entonces

\begin{equation}
f_a(t) \rightarrow \frac{1}{|a|} F(\frac{u}{a})    \;\;\;\;\; \mbox{o, equivalentemente}\;\;\;\;\; \boxed{f(at) \rightarrow  \frac{1}{|a|}F(\frac{u}{a})}
\end{equation}

Vemos que la Transformada de Fourier intercambia la dilatación con la compresión, y viceversa.


### Derivada

En general, si $f  \in   L^1(\mathbb{R};\mathbb{C}) $, su derivada no es necesariamente integrable. Similarmente, su Transformada de Fourier $F(u)$ no es necesariamente derivable. Hay dos condiciones necesarias bajos las cuales es posible definir la Transformada de Fourier de $df/dt$ y bajo las cuales la derivada $dF/du$ existe, y veremos que en este caso la Transformada de Fourier intercambia la multiplicación por $t$ (o $u$) con la derivada respecto a $u$ (o $t$).


- **Derivación en frecuencia**. Si $ t f  \in   L^1(\mathbb{R};\mathbb{C}) $, entonces su Transformada de Fourier tiene una derivada continua, $F(u) \in C^1(\mathbb{R};\mathbb{C})$ y tenemos:

\begin{equation}
\frac{d F(u)}{du} = \int_{\mathbb{R}} e^{-i 2 \pi u t} (-i 2 \pi t) f(t) dt = \mathcal{F}\{-i 2 \pi t f\} (u) 
\end{equation}

Otra forma de escribir esta condición es

$$ \boxed{t f(t) \rightarrow \frac{i}{2 \pi} F'(u)} $$

- **Derivación en el tiempo o espacio**. Si $f \in C^1(\mathbb{R};\mathbb{C})$ y $f' \in  L^1(\mathbb{R};\mathbb{C})$, entonces:

\begin{equation}
\mathcal{F}\{\frac{d  f}{dt}\}(u) = \int_{\mathbb{R}} e^{-i 2 \pi u t} \frac{d f(t)}{dt} dt = i 2 \pi u \mathcal{F}\{f\} (u)
\end{equation}

o en forma más sucinta:

$$ \boxed{f'(t) \rightarrow i 2 \pi u F(u)} $$

Vemos entonces que bajo estas condiciones, una multiplicación por $-i 2 \pi t$ corresponde a una derivada en el dominio de Fourier $u$, mientras que una derivada respecto a $t$ corresponde a una multiplicación por $i 2 \pi u$ en el dominio de Fourier. En consecuencia, la multiplicación por la variable independiente y la derivada son intercambiadas por la Transformada de Fourier.


### Convolución

La convolución $g = h \ast f  \in  L^1(\mathbb{R};\mathbb{C})  $ y su Transformada de Fourier está dada por:
 
 $$ G(u) = \mathcal{F}\{ h \ast f\} (u) = H(u) F(u) $$
 
 
En el caso en que $h(t)$ represente la respuesta al impulso de un sistema, $f(t)$ la entrada y $g(t)$ la salida, podemos ver que su Transformada de Fourier $H(u)$ se encuentra simplemente mediante la expresión:

$$ \boxed{H(u) = \frac{G(u)}{F(u)}} $$

<div class="alert alert-block alert-info">
    $H(u)$ se conoce como <b>respuesta de frecuencia</b>. Esta propiedad implica que el cálculo de la salida de un sistema LTI a una entrada arbitraria implica una simple multiplicación en el dominio de Fourier.
</div> 

### Multiplicación

El producto de dos funciones $fg$ no siempre es integrable, por lo que no es posible definir de forma general $\mathcal{F}\{fg\}$. Pero si $\mathcal{F}\{g\} \in L^1(\mathbb{R};\mathbb{C})$, entonces su transformada inversa es  $g$ lo que implica que es una función acotada y continua, $g \in C_b(\mathbb{R};\mathbb{C})$, lo que asegura que el producto $fg$ $\in   L^1(\mathbb{R};\mathbb{C})$

Por otro lado, la convolución de  $\mathcal{F}\{f\}$ y  $\mathcal{F}\{g\}$ está bien definida dado que $\mathcal{F}\{g\} \in L^1(\mathbb{R};\mathbb{C})$ y $\mathcal{F}\{f\} \in C_b(\mathbb{R};\mathbb{C})$ Entonces:


$$  \mathcal{F}\{fg\} (u) = \int_{\mathbb{R}} f(t) e^{-i 2 \pi u t} \left\{  \int_{\mathbb{R}} e^{i 2 \pi t \nu} \mathcal{F}\{g\} (\nu) d\nu  \right\} dt  $$

donde $g$ está escrita con la formula inversa. Como $| f(t) \mathcal{F}\{g(\nu)\} | $ es integrable sobre $\mathbb{R}^2$, el Teorema de Fubini nos da:
$$  \mathcal{F}\{fg\}(u) = \int_{\mathbb{R}}   \left\{ \int_{\mathbb{R}} f(t)   e^{-i 2 \pi t(u- \nu)} dt \right\} \mathcal{F}\{g\} (\nu) d\nu   $$

Aquí reconocemos la Transformada de Fourier de $f$ evaluada en $u-\nu$


\begin{equation}
  \mathcal{F}\{fg\}(u) = \int_{\mathbb{R}}   \mathcal{F}\{f\}(u-\nu) \mathcal{F}\{g\}(\nu) d\nu  =  \mathcal{F}\{f\} \ast \mathcal{F}\{g\} (u) 
  \end{equation}

En forma abreviada tenemos

$$ \boxed{\{f g\}(t) \rightarrow \{F \ast G\}(u)}$$

Vemos que, bajo ciertas condiciones, la Transformada de Fourier intercambia la multiplicación con la convolución.

### Fórmula de Transferencia

Esta fórmula juega un rol clave para poder definir la Transformada de Fourier a una clase más extendida de objetos tales como las distribuciones. Dice así:


\begin{equation}
\boxed{ \int_{\mathbb{R}}   \mathcal{F}\{f\} (t) g(t) dt = \int_{\mathbb{R}}   f(t) \mathcal{F}\{g\}(t) dt}  
\end{equation}


### Plancherel-Parseval
El teorema de Plancherel-Parseval dice que:

Si $f,g \in   L^1(\mathbb{R};\mathbb{C}) \cap L^2(\mathbb{R};\mathbb{C})$, entonces sus Transformadas de Fourier están en $L^2(\mathbb{R};\mathbb{C}) $ y se tiene que:

\begin{equation}
\boxed{(\mathcal{F}\{f\}, \mathcal{F}\{g\} ) = (f,g ); \; \;\;\;\;\; \| \mathcal{F}\{f\} \|_{L^2}^2  = \|  f \|_{L^2}^2   }  
\end{equation}
Esto dice que la Transformada de Fourier de funciones que son integrables y cuadrado integrables conservan el producto interno y la norma $L^2$.

Otra forma de escribir esta propiedad, considerando el producto interno de una función consigo misma, es:

$$ (f, f) = \int_{\mathbb{R}} |f(t)|^2 dt = \int_{\mathbb{R}} |F(u)|^2 du $$

## La Transformada de Fourier intercambia suavidad por decaimiento

Sea $f \in  L^1(\mathbb{R};\mathbb{C})  $ de la clase $C^k (\mathbb{R};\mathbb{C})$ (todas las derivadas de orden $\alpha \leq k$ existen y son continuas). Si $  \forall \alpha \leq k $, $ \frac{d^\alpha f}{dt^\alpha} \in L^1(\mathbb{R};\mathbb{C})$, entonces

$$ \mathcal{F}\{ \frac{d^{\alpha}f}{dt^\alpha}\} = (i 2\pi u)^\alpha \mathcal{F}\{f\}(u)$$

y $\mathcal{F}\{f\}(u) = O(1/|u| ^k )$ para $| u  |\rightarrow \infty $.

Es más, esto puede ser demostrado al aplicar $k$ veces la derivación en el tiempo de la propiedad de la diferenciación. 

Dado que para todo $\alpha \leq k$, $(i 2 \pi u)^\alpha \mathcal{F}\{f\}$ es la Transformada de Fourier de una función integrable, el Teorema de Riemann-Lebesgue asegura que $\forall \alpha \leq k$, $ | u  |^\alpha | \mathcal{F}\{f\}(u)  | \rightarrow 0$, y sigue que $\mathcal{F}\{f\}$ decae como $1/u^k$.

Entonces, mientras más suave sea la función $f$ (mayor el valor de $k$), más rápido la Transformada de Fourier va a decaer en $| u | \rightarrow  \infty$.

Si $f$ es tal que $(1+ t^k)f \in L^1(\mathbb{R};\mathbb{C}) $, entonces 
 $\mathcal{F}\{f\}$ es de clase $C^k (\mathbb{R};\mathbb{C})$ y

$$ \frac{d^{\alpha}\mathcal{F}\{f\}}{du^\alpha} = (-i 2 \pi)^\alpha \mathcal{F}\{ t^\alpha f \} \;\;\;\;  \forall \alpha \leq k, \alpha \in \mathbb{N} $$

Mientras más rápido decaiga la función $f$ en $| t  |\rightarrow \infty$, más suave será su Transformada de Fourier. Esto se puede demostrar aplicando $k$ veces la propiedad de la diferenciación en frecuencia, y deducimos que todas las derivadas de $\mathcal{F}\{f\}$ hasta el orden $\alpha \leq k$ son Transformadas de Fourier de funciones integrables, y por lo tanto continuas dado el Teorema de Riemann-Lebesgue.

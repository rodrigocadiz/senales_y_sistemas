## Motivación

En el análisis de Fourier, las bases son exponenciales complejas en función de una variable real $\omega$ o $u$, la cual puede ser interpretada como frecuencia y la Transformada de Fourier puede, a su vez, ser concebida como un espectro. Sin embargo, como hemos visto en estos apuntes, la convergencia de la integral para funciones fuera de $L^1$ no está garantizada. Una forma posible de subsanar este obstáculo es considerar una variable compleja $s$ en vez de la real $\omega$ o $u$ en los exponentes de las bases de Fourier, donde $s=\sigma+i\omega = \sigma + i 2 \pi u$.

Este cambio de variables implica que las bases originales de Fourier serían ahora:

$$ e^{-st} = e^{-\sigma t}e^{-i\omega} = e^{-\sigma t}e^{-i 2 \pi u}$$

Para ciertos valores de $\sigma$, el término $e^{-\sigma t}$ fuerza a una función a decaer exponencialmente a cero en $t \rightarrow \infty$.  Esta idea da origen a la Transformada de Laplace, la cual se puede aplicar en forma general a funciones localmente integrables, y que no son necesariamente integrables sobre $\mathbb{R}$.  

<div class="alert alert-block alert-info">
Todas la funciones que tienen una Transformada de Fourier en $L^1(\mathbb{R};\mathbb{C})$ admiten una Transformada de Laplace, pero no viceversa.
</div>

## Transformada de Laplace

La [Transformada de Laplace](https://en.wikipedia.org/wiki/Laplace_transform) es una transformada integral utilizando una exponencial compleja $e^{s t}$ como las bases de la transformación. Mapea una señal compleja  $f(t) \in \mathbb{C}$ con una variable independiente real $t \in \mathbb{R}$ hacia una transformada compleja $F(s) \in \mathbb{C}$ con una variable independiente compleja $s \in \mathbb{C}$. Esta transformada debe su nombre a [Pierre-Simon Laplace](https://en.wikipedia.org/wiki/Pierre-Simon_Laplace).

<div class="alert alert-block alert-info">
Como veremos más adelante, las exponenciales complejas son autofunciones de los sistemas LTI lineales e invariantes, lo que hace a la transformada de Laplace una herramienta muy importante para el análisis de estos sistemas.
</div>

### Transformada de Laplace (unilateral)

Dado que las señales causales juegan un rol muy importante en la teoría de señales y sistemas, por lo general la Transformada de Laplace se utiliza en este tipo de señales. Sea $f: \mathbb{R} \mapsto \mathbb{C}$ una función causal, esto es, $f=0$ for $t<0$.
La Transformada de Laplace $\mathcal{L}\{f\}$ de $f$ es una aplicación $\mathbb{C} \mapsto \mathbb{C}$  definida por:

\begin{equation}
\boxed{\mathcal{L}\{f\}(s) = F(s) = \int_{0}^{\infty} f(t) \, e^{- s t} \; dt}
\end{equation}

donde $F(s) = \mathcal{L} \{ f(t) \}$ denota la Transformada de Laplace de $f(t)$. A veces, para diferenciarla de la Transformada bilateral usaremos la notacion $\mathcal{L}_1\{\}$ o bien $F_1(s)$.


Una señal compleja $f(t) \in \mathbb{C}$, función de la variable independiente $t \in \mathbb{R}$ se representa por su Transformada de valor complejo $F(s) \in \mathbb{C}$, función de una variable independiente compleja $s \in \mathbb{C}$. La variable $s$ se puede interpretar como una [frecuencia compleja](../continuous_signals/standard_signals.ipynb#Complex-Exponential-Signal), donde $s = \sigma + i \omega = \sigma + i 2 \pi u$ con $\sigma, \omega, u \in \mathbb{R}$.

<div class="alert alert-block alert-info">
Es claro que la integral antes mencionada no siempre existe. Su existencia depende de tanto de la frecuencia compleja $s$ como de la señal $f(t)$. Todos los valores de $s$ para los cuales la Transformada de Laplace converge se denomina <a href="https://en.wikipedia.org/wiki/Radius_of_convergence"><b>región de convergencia</b></a> o simplemente ROC. Las transformadas Laplace de dos señales distintas pueden diferir solamente respecto a sus ROCs. Consecuentemente, no es posible invertir esta transformada sin considerar el valor de la ROC.
</div>

Definamos la absica mínima como el menor valor de $\sigma \in \mathbb{R}$ tal que $f(t)e^{-\sigma t}$ es integrable:

$$ \alpha = \inf \{  \sigma \in \mathbb{R} \;\;|\;\;  t \mapsto f(t)e^{-\sigma t} \in L^1(\mathbb{R}^+;\mathbb{C})   \}  $$

entonces, la Transformada de Laplace de $f$ se define para $\text{Re}\{s\} > \alpha$. Para $\text{Re}\{s\}=\alpha$, la definición integral anterior puede o no converger.

La Transformada de Laplace se usa ampliamente para problemas de valores iniciales, tales como transientes en sistemas dinámicos, pero no tiene una interpretación física tan clara como la Transformada de Fourier. Observamos que la existencia de $\mathcal{L}\{f\}(s)$ impone que:

$$ t\mapsto |f(t)e^{-st}| = |f(t)| e^{-\text{Re}\{s\} t } \;\;\; \in L^1(\mathbb{R}^{+};\mathbb{C}) $$

### Transformada de Laplace bilateral

La Transformada de Laplace bilateral se define como:

\begin{equation}
\mathcal{L}\{f\}(s) = F(s) = \int_{-\infty}^{\infty} f(t) \, e^{- s t} \; dt
\end{equation}

Obviamente, la transformada bilateral entrega el mismo resultado que la transformada unilateral para señales causales. A veces, para diferenciarla de la Transformada unilateral usaremos la notacion $\mathcal{L}_2\{\}$ o bien $F_2(s)$.

La relación entre ambas transformadas es

$$\boxed{L_2\{f(t)u(t)\} = L_1\{f(t)\}}$$

donde $u(t)$ es el escalón.

## Convergencia y ROCs
Una condición necesaria pero no suficiente para la existencia de una Transformada de Laplace bilateral $F(s) = \mathcal{L} \{ f(t) \}$ de una señal $f(t)$ es

\begin{equation}
\int_{-\infty}^{\infty} | f(t) \cdot e^{- \sigma t} | \; dt < \infty
\end{equation}

donde $\sigma = \Re\{ s \}$ es la parte real de la frecuencia compleja $s = \sigma + i \omega$ de la Transformada de Laplace. La señal $f(t)$ debe ser [localmente integrable](https://en.wikipedia.org/wiki/Locally_integrable_function).

La parte imaginaria $\omega = \Im \{ s \}$ de la frecuencia compleja $s$ no tiene efecto en la convergencia de la integral. Consecuentemente, la ROC se determina solamente por su parte real $\sigma = \Re\{ s \}$. Para una señal derecha (causal) con $f(t) = 0$ para $t < T_0$, la ROC está dada por su absisa mínima de converegencia $\Re \{ s \} > a$ con $a \in \mathbb{R}$. Esto se puede concluir del decaimiento del término exponencial  $e^{- \sigma t}$ para $t \to \infty$, lo que asegura la convergencia para un $a$ dado. Sin embargo, este razonamiento solo es aplicable a señales con [crecimiento exponencial](https://en.wikipedia.org/wiki/Exponential_growth). 

El mismo razonamiento lleva, en el caso de la Transformada bilateral a la determinación de las ROCs para señales izquierdas (anti-causales) o bilaterales. Las ROCs resultantes se resumen en la siguiente ilustración

![Regiones de convergencia para señalas derechas/bilaterales/izquierdas](./images/ROC.png)

Las areas grises son los valores de $s$ para los cuales la Transformada de Laplace converge. Los bordes $a$ de esas áreas (dashed lines) dependen de la señal $f(t)$. 


### Resumen
<div class="alert alert-block alert-info">
<ul>
<li>Si $F(s)$ es la Transformada de Laplace de una función y es racional, entonces su ROC está limitada por sus polos o bien se extiende hacia el infinito.
<li>La ROC no puede contener ningún polo.
<li>Si $F(s)$ es la Transformada de Laplace racional de una función derecha, la ROC es la región hacia la derecha del polo más a la derecha.
<li>Si $F(s)$ es la Transformada de Laplace racional de una función izquierda, la ROC es la región hacia la izquierda del polo más a la izquierda.
<li>Si $F(s)$ es la Transformada de Laplace racional de una función bilateral, la ROC es la región determinada por sus polos interiores.
</ul>
</div>

## Representación en el plano $s$

La transformada de Laplace $F(s) = \mathcal{L} \{f (t) \} $ depende de la frecuencia compleja $s \in \mathbb {C} $ y en general tiene un valor complejo $ F(s) \in \mathbb {C} $. Se puede graficar considerando su magnitud $ | F (s)| $ y fase $ \phi(s) $ o partes real $\Re \{F(s) \} $ e imaginaria $ \Im \{F(s) \} $ en el plano complejo $s$. 

Sin embargo, muchas transformaciones de Laplace de interés en la teoría de señales y sistemas son [funciones racionales](https://en.wikipedia.org/wiki/Rational_function) en $s$. Los polinomios del numerador y el denominador se pueden representar por sus [raíces complejas](https://en.wikipedia.org/wiki/Zero_of_a_function#Polynomial_roots) y un factor constante. Las raíces del numerador se denominan *ceros*, mientras que las raíces del denominador se denominan *polos* de $F(s)$. El polinomio y la representación en polos y ceros de una transformada racional de Laplace son equivalentes

\begin{equation}
F(s) = \frac{\sum_ {m = 0} ^ {M} \beta_m s^m}{\sum_{n = 0} ^ {N} \alpha_n s^n} = K \cdot \frac{\prod_{\mu = 0}^{Q}(s - s_{0 \mu})}{\prod_{\nu = 0}^{P}(s - s_{\infty \nu})}
\end{equation}

donde $M$ y $N$ denotan el orden del polinomio numerador y denominador, $ s_{0 \mu} $ y $ s_{\infty \nu} $ los $ \mu$-ésimi cero / $\nu$-ésimo polo de $F(s) $ y $ Q = M-1 $ y $ P = N-1 $ el número total de ceros y polos, respectivamente. Para $ M = N $, el factor $ K = \frac {\beta_M} {\alpha_N} $. Si $ M> N $, al menos un polo se encuentra en $ |s| = \infty $.

Es común ilustrar los polos y ceros en un [diagrama de polos y ceros](https://en.wikipedia.org/wiki/Pole%E2%80%93zero_plot). Aquí las ubicaciones de los polos y ceros complejos, sus grados y el factor $ K $ se dibujan en el plano $s$. Los polos se denotan con "x" y los ceros con "o".

![Ubicaciones de los polos y cero en un diagrama de polos cero](./images/splane.png)

Las ubicaciones de los polos y ceros proporcionan información sobre la composición de una señal. Por ejemplo, la [transformada de Laplace de una señal exponencial](#exponencial) tiene un polo en $s = -a$. Una señal que se compone de una superposición de señales exponenciales tendrá múltiples polos cuyas posiciones están relacionadas con las frecuencias complejas de las que está compuesta.

### Ejemplo: Transformada de Laplace del escalón

Consideremos la función escalón $u(t)$. Su absica mínima de convergencia es $\alpha =0$, dado que $t \mapsto u(t)e^{-\sigma t} $ es integrable sobre $\mathbb{R}^+$ para $\sigma > 0$. Su Transformada de Laplace está dada por:

$$  \mathcal{L}\{u(t)\}(s) = \int_{\mathbb{R}^+}e^{-st}dt = \frac{1}{s}\;\;\;\; \text{Re}\{s\} > 0  $$

### Ejemplo: Transformada de Laplace de una potencia de $t$

Consideremos la función $f:\mathbb{R}^+ \rightarrow (0,\infty)$ dada por 

$$f(t) = t^n \;\;\;\; n \in \mathbb{N} $$

Este es un ejemplo típico de una función no integrable sobre $\mathbb{R}$ cuya Transformada de Fourier no existe fuera del marco de las distribuciones. Su Transformada de Laplace, sin embargo, está bien definida para  $\text{Re}\{s\} > 0$. En efecto, $t \rightarrow te^{-\sigma t}$ es integrable sobre $\mathbb{R}$ para $\sigma >0$, por lo que la absica mínima de convergencia para $f$ es $\alpha =0$. Veamos el caso simple $n=1$ ($f(t)=t$). Integrando por partes, obtenemos:

$$\mathcal{L}\{ t\} = \int_{\mathbb{R}^+} t e^{-st}dt = \frac{1}{s} \int_{\mathbb{R}^+}e^{-st}dt = \frac{1}{s^2}\;\;\;  \text{Re}\{s\} > 0 $$

y por recurrencia, obtenemos para el caso genérico $f(t)= t^n$, $n \in \mathbb{N}$

$$\mathcal{L}\{ t^n\} = \int_{\mathbb{R}^+} t^n e^{-st}dt = \frac{n}{s} \int_{\mathbb{R}^+}t^{n-1}e^{-st}dt = \frac{n!}{s^{n+1}}\;\;\;  \text{Re}\{s\} > 0 $$

El caso particular $n=0$ corresponde a la función escalón cuya Transformada de Laplace es, en efecto, $1/s$. 

<a name="exponencial"></a>
### Ejemplo: Transformada de Laplace de una exponencial 
Consideremos una función exponencial $t \rightarrow e^{at}$ con $a \in \mathbb{R}$. Esta función no es integrable para $a>0$. Su Transformada de Laplace está bien definida para $\text{Re}\{s \} > a$:

$$  \mathcal{L}\{ e^{at} \}(s) = \int_{\mathbb{R}^+} e^{at} e^{-st}dt = \frac{1}{s-a} \;\;\;  \text{Re}\{s\} > a $$


### Ejemplo: Transformada de Laplace de $1/\sqrt{t}$
Consideremos la función $f:\mathbb{R}^{+} \rightarrow (0,\infty)$ dada por

$$ f(t) = \frac{1}{\sqrt{t}} \;\;\;\; t >0 $$

Su Transformada de Laplace existe para $\text{Re}\{s \} > 0$ dado que $t^{-1/2}e^{-\sigma t} \sim e^{-\sigma t}$ cuando $t \rightarrow \infty$. Tenemos

$$  \mathcal{L}\{ \frac{1}{\sqrt{t}} \}(s) = \int_{\mathbb{R}^+} \frac{1}{\sqrt{t}} e^{-st}dt = 2\int_{\mathbb{R}^+} e^{-su^2}du=\sqrt{\frac{\pi}{s}} \;\;\;  \text{Re}\{s\} > 0 $$

donde hemos usado


$$  2\int_{\mathbb{R}^+} e^{-su^2}du = \int_{\mathbb{R}} e^{-su^2}du  = I $$

y

$$ I^2 = \int_{\mathbb{R}} \int_{\mathbb{R}}e^{-s(x^2+y^2)}dxdy = 2\pi \int_{0}^{\infty} e^{-s r^2}r dr =  \frac{\pi}{s}  $$

## Propiedades de la Transformada de Laplace

Al estudiar las propiedades de la Transformada de Laplace, el paralelo con muchas de las propiedades de la Transformada de Fourier son evidentes. Por ejemplo, un desplazamiento de $f$ corresponde a una multiplicación de $\mathcal{L}\{f\}$ por una exponencial. Para estas propiedades consideraremos funciones derechas y ROC genérica $\sigma > \alpha $.

### Linealidad

La Transformada de Laplace es una aplicación lineal, por lo que cumple con homogeneidad, para todo $a \in \mathbb{R}$

$$ \boxed{a f(t) \rightarrow a F(s)} $$ 

y superposición

$$ \boxed{f(t)+g(t) \rightarrow F(s)+G(s)} $$

con ROC $\max(\alpha_f,\alpha_g) < \sigma $

### Desplazamiento
Sea $f:\mathbb{R}  \mapsto  \mathbb{C}$ una función causal y $\mathcal{L}\{ f\}$ su Transformada de Laplace. Entonces, para todo $a \in \mathbb{R}^+$ y $b \in \mathbb{C}$

\begin{equation}
\mathcal{L}\{ f(t-a) u(t-a)\}(s) = e^{-as} \mathcal{L}\{f\}(s)  \;\;\;\;\; \mbox{y}\;\;\;\;\;  \mathcal{L}\{e^{-bt}f(t)\} =  \mathcal{L}\{f\}(s+b)  
\end{equation}

con ROC $\alpha < \sigma < \beta $

#### Ejemplo:
Consideremos ahora $t\mapsto t^n e^{at}$. Su Transformada de Laplace está dada por

$$ \mathcal{L}\{ t^n e^{at}  \}(s)  =  \mathcal{L}\{ t^n   \} (s-a) = \frac{n!}{(s-a)^{n+1}} $$

similarmente, la Transformada de Laplace de $t \mapsto e^{-at}\sin at $ se puede obtener

$$ \mathcal{L}\{ e^{-at}\sin at  \}(s)  =  \mathcal{L}\{ \sin at   \} (s+a) = \frac{a}{(s+a)^2 + a^2} $$

#### Ejemplo: Transformada de Laplace de $e^{\pm it}/\sqrt{t} u(t)$

Recordando que 
$$ \mathcal{L}\{ \frac{H(t)}{\sqrt{t}}  \}(s)  = \sqrt{\frac{\pi}{s}}\;\;\;\; \text{Re}\{s\} > 0$$

la propiedad del desplazamiento da
$$ \mathcal{L}\{ e^{\pm it}\frac{H(t)}{\sqrt{t}}  \}(s)  = \mathcal{L}\{ \frac{H(t)}{\sqrt{t}}  \}(s\mp i) = \sqrt{\frac{\pi}{s \mp i}}\;\;\;\; \text{Re}\{s\} > 0$$

### Escalamiento
Sea $f:\mathbb{R}  \mapsto  \mathbb{C}$ una función causal y $\mathcal{L}\{ f\}$ su Transformada de Laplace. Para $a \in \mathbb{R}/\left\{0\right\}$ definimos $ f_a(t) = f(at) $. Entonces

\begin{equation}
\mathcal{L}\{  f_a (t) \}(s) = \mathcal{L}\{  f(at) \}(s)= \frac{1}{|a|}\mathcal{L}\{  f(t) \}(\frac{s}{a})
\end{equation}

con ROC $\alpha < \frac{\sigma}{a} $


### Derivada 

- Supongamos una función causal $f$ con absisa mínima de converegencia $\alpha$. Entonces, la función $t \in \mathbb{R}^+ \mapsto (-t)^m f(t)$ tiene la misma absisa mínima de converegencia. Más aún, la Transformada de Laplace $\mathcal{L}\{f\}$ es [holomórfica](https://en.wikipedia.org/wiki/Holomorphic_function) en el plano $\text{Re}\{s\} > \alpha $ y

\begin{equation}
\frac{d^m \mathcal{L}\{f\} (s)}{d s^m} = \int_{\mathbb{R}^+}  (-t)^m f(t)e^{-s t} dt = \mathcal{L}\{(-t)^m f\} (s)
\end{equation}

En efecto, la derivada bajo el signo de la integral se justifica dado que para todo  $s$ tal que $\text{Re}\{s\} > \alpha$ 

$$  |(-t)^m f(t) e^{-st}| \leq t^m|f(t)|e^{-\alpha t} \in L^1(\mathbb{R},\mathbb{C})  $$

- Sea $f$ una función de clase $C^1(\mathbb{R};\mathbb{C})$ cuya Transformada de Laplace tiene una absisa mínima de converegencia  $\alpha$. Si además para $\text{Re}\{s\} > \alpha$ la función $t \mapsto e^{-st}f(t)$ admite un límite cuando $t \rightarrow \infty$, entonces la función $f'$ tiene una Transformada de Laplace con la misma absisa mínima $\alpha$ y

\begin{equation}
\mathcal{L}\{ f'(t) \}(s) = s \mathcal{L}\{f\}(s) - f(0^+) 
\end{equation}

Este resultado se puede generalizar para una derivada de orden $n$, donde la continuidad de las primeras $(n-1)$ derivadas en $t \neq 0$ se requiere:

\begin{equation}
\mathcal{L}\{ \frac{d^n f}{dt^n} \}(s) = s^n \mathcal{L}\{f\}(s) -s^{n-1}f(0^+)  -s^{n-2}\frac{df}{ds}(0^+)  - ... - \frac{d^{n-1}f}{ds^{n-1}}(0^+) 
\end{equation}

#### Ejemplo: Transformada de Laplace de $te^{-at}$ y $t\sin t$
Consideremos la función $f(t) = te^{-at}$. Su Transformada de Laplace es


$$\mathcal{L}\{t e^{-at}\} (s) = -\frac{d}{ds}\mathcal{L}\{e^{-at}\} (s) = -\frac{d}{ds}\left( \frac{1}{s+a}\right) \;\;\;\; \text{Re}\{ s\} > a $$

En la misma forma, la Transformada de Laplace de $g(t) = t\sin t$ está dada por

$$\mathcal{L}\{t \sin t \} (s) = -\frac{d}{ds}\mathcal{L}\{  \frac{1}{s^2+1} \} (s) = \frac{2s}{(s^2+1)^2} \;\;\;\; \text{Re}\{ s\} > a $$

#### Ejemplo: Transformada de Laplace de $\sin t u(t)$
Consideremos la función causal $f(t)= \sin t u(t)$. Su Transformada de Laplace es fácilmente obtenible. Dado que $f''(t)= -\sin t u(t) = -f(t)$

$$\mathcal{L}\{f''(t)\}(s)= - \mathcal{L}\{f\}(s)=  s^2 \mathcal{L}\{f\}(s) - s f(0) - \underbrace{f'(0)}_{=1}  $$ 

entonces

$$ \mathcal{L}\{ \sin t u(t) \} = \frac{1}{s^2+1} \;\;\;\; \text{Re}\{s\} >0  $$


### Integración 
Sea $g(t)=\int_0^t f(t')dt'$ la primitiva de $f$ la cual desaparece en $t=0$, entonces


\begin{equation}
\mathcal{L}\{ \int_0^t f(t')dt' \}(s) = \frac{ \mathcal{L}\{f\}(s)  }{s}
\end{equation}



### Convolución
Supongamos dos funciones causales $f $ y $g$. Entonces la convolución $ (f \ast g )(t) = \int_0^t f(t')g(t-t')dt' $ es también causal y su Transformada de Laplace está dada por:

\begin{equation}
\mathcal{L}\{ f \ast g \}(s) =  \mathcal{L}\{f\}(s) \mathcal{L}\{g\}(s) \;\;\;\; \text{Re}\{s\} > \alpha
\end{equation}

donde la ROC es $\max(\alpha_f,\alpha_g) < \sigma  $


#### Ejemplo: Transformada de Laplace de $tu(t) \ast t^2u(t)$

Consideremos la convolución $g$ entre las funciones causales $f(t) = t u(t)$ and $v(t)=t^2u(t)$. Está dada por


$$ g(t) = \int_0^{\infty}   f(t') v(t-t') dt' =  \int_0^{t}  t' (t-t')^2 dt' = \frac{t^4}{12} u(t)  $$

su Transformada de Laplace se puede obtener

$$  \mathcal{L}\{\frac{t^4}{12}u(t) \}(s) = \frac{4!}{12 s^5}=\frac{2}{s^5} \;\;\;\; \text{Re}\{s\} > 0  $$

El mismo resultado se puede obtener rápidamente usando la propiedad de la convolución

$$  \mathcal{L}\{\frac{t^4}{12}u(t) \}(s)  =  \mathcal{L}\{  tu(t) \ast t^2u(t)  \}(s)=   \mathcal{L}\{  tu(t)  \}(s) \mathcal{L}\{   t^2u(t)  \}(s)$$ $$=\frac{1}{s^2} \frac{2!}{s^3} = \frac{2}{s^5}\;\;\; \text{Re}\{s\} > 0  $$

## Comportamiento asintótico
Sea $f$ una función causal cuya Transformada de Laplace $\mathcal{L}\{f\}$ tiene una absisa mínima de convergencia $\alpha$. Entonces, para $\text{Re}\{s\} > \alpha$ $\mathcal{L}\{f\}$ es holomórfica y acotada. Además, tiene a cero en el infinito


Sea $f$ una función con absica mínima de convergencia $\alpha$. Entonces

\begin{equation}
\lim_{|s| \rightarrow \infty} \mathcal{L}\{f\}(s) = 0 \;\;\;\; \text{Re}\{s\} > \alpha
\end{equation}


Observación: La función $z\in\mathbb{C} \mapsto 1$ no puede ser la Transformada de Laplace de ninguna función, dado que no decae a cero en el infinito. Veremos, sin embargo, que al extender la Transformada de Laplace a las distribuciones, es posible mostrar que $\mathcal{L}\{\bf{\delta}\} = \bf{1}$.

## Teorema del valor final
Tal como es el caso de la Transformada de Fourier en $L^1$, hay una correspondencia entre el comportamiento de una función $f$ en el infinito (o en $t=0$) y el de su Transformada de Laplace en $|s|=0$ (o $|s|=+\infty)$. 

Sea $f $ una función causal tal que $f$ y $f'$ tienen Transformadas de Laplace. Si $f$ tiene un límite $f(+\infty)$ cuando $t \rightarrow +\infty$, entonces $\mathcal{L}\{f\}$ verifica

$$\lim_{|s|\rightarrow 0^+} s \mathcal{L}\{f\}(s) = f(+\infty)  $$



## Teorema del valor inicial 
Sea $f $ una función causal tal que $f$ and $f'$ admiten Transformadas de Laplace. Si $f$ tiene un límite $f(0^+)$ cuando $t\rightarrow 0$, entonces $\mathcal{L}\{f\}$ verifica

$$\lim_{|s|\rightarrow +\infty} s \mathcal{L}\{f\}(s) = f(0^+)  $$

### Ejemplo: Valores asintóticos del escalón
Para el escalón $u(t)$ se tiene:

$$  u(0^+) = u(+\infty) = 1  $$

Su Transformada de Lapalce $\mathcal{L}\{u\}(s) = \frac{1}{s} \;\; \text{Re}\{s\} >0 $ y verifica

$$  \lim_{|s|\rightarrow +\infty} s \mathcal{L}\{u\}(s) = 1 =   u(0^+) $$
y
$$  \lim_{|s|\rightarrow 0^+} s \mathcal{L}\{u\}(s) = 1 =   u(+\infty) $$

### Ejemplo: Valores asintóticos de $\cos t$
Para la función $f(t)=\cos(t)$ se tiene:

$$  f(0^+) = 1  $$

Su Transformada de Laplace $\mathcal{L}\{\cos t\}(s) = \frac{s}{s^2+1}$ y, en efecto,

$$  \lim_{|s|\rightarrow +\infty} s \mathcal{L}\{\cos(t)\}(s) = 1 =   f(0^+)  $$

pero además, $f(+\infty)$ no existe, por lo que

$$  \lim_{|s|\rightarrow 0^+} s \mathcal{L}\{\cos (t)\}(s) = 0 $$

## Transformada de Laplace de una distribución
Recordemos que el soporte $\text{supp}(\bf{T})$ de una distribución $\mathbf{T}$ es el conjunto cerrado más pequeño fuera del cual la distribución es cero. Notamos que $\mathcal{D}_{+}^{'}(\mathbb{R})$ es el conjunto de distribuciones cuyo soporte está incluído en $[0,+\infty[$.

La Transformada de Laplace de una distribución $\bf{T}\in \mathcal{D}_{+}^{'}(\mathbb{R})  $ con soporte compacto es la función $s \in \mathbb{C} \rightarrow T(s)$ definida por

\begin{equation}
T(s) = \langle \mathbf{T}, e^{-st}\rangle
\end{equation}

- En el caso de una distribución regular, la definición coincide con la Transformada de Laplace para funciones.

- La función $s \mapsto e^{-st}$ es $C^{\infty}(\mathbb{C};\mathbb{C})$ pero no tiene soporte compacto. $T(s)$ está bien definida porque la distribución $\mathbf{T}$ en sí misma tiene soporte compacto.
 
***
#### Ejemplo: Transformada de Laplace del impulso
Consideremos la distribución de Dirac $\mathbf{\delta}$, cuyo soporte es $\text{supp}(\mathbf{\delta})=\{0\} \in [0,\infty)$. Su Transformada de Laplace $\mathcal{L}\{ \mathbf{\delta} \}(s)$ es 

$$ \mathcal{L}\{ \mathbf{\delta} \}(s)= \langle \mathbf{\delta}, e^{-st} \rangle = 1  $$

Veamos como en **sympy** se puede calcular esta Transformada:

```python
%matplotlib inline
import sympy as sym
sym.init_printing()

t = sym.symbols('t', real=True)
s = sym.symbols('s', complex=True)

X = sym.integrate(sym.DiracDelta(t)*sym.exp(-s*t), (t, -sym.oo, sym.oo))
X
```

### Propiedad de la derivada en distribuciones 
Sea $\mathbf{T}$ una distribución que admite una Transformada de Laplace. La derivada $D^{1}\mathbf{T}$ admite una Transformada de Laplace que verifica:

\begin{equation}
\mathcal{L}\{ D^{1}\mathbf{T} \}(s) = s T(s)
\end{equation}

la función $s \mapsto T(s)$ es holomórfica en todo el plano complejo y se tiene 

\begin{equation}
T'(s) = \langle -t \mathbf{T}    ,e^{-st}\rangle \label{Tl_t_distr}
\end{equation}

La generalización a un orden $n$ es:
\begin{equation}
\mathcal{L}\{ D^{n}\mathbf{T} \}(s) = s^n T(s) \;\;\;\;\; \frac{d^n T(s)}{ds^n} = \langle (-t)^n \mathbf{T}    ,e^{-st}\rangle 
\end{equation}


- Esta propiedad es similar para el [caso de funciones causales](), excepto que para la contribución de $f(0^+)$ que ya ha sido tomada en consideración en el contexto de las distribuciones. En efecto, sea $f=gu$ una función causal donde $g \in C^1(\mathbb{R},\mathbb{C})$ y $u$ es el escalón. Esto define una distribución regular $\mathbf{f}$ tal que

$$ \mathcal{L}\{D^{1}\mathbf{f}\} (s)= s  \mathcal{L}\{ \mathbf{f}\} (s)  $$
 
en la otra mano, la formula de saltos da
 
$$  D^{1}\mathbf{f} =  \mathbf{f'} + f(+0)\mathbf{\delta} $$
 
tal que 

$$ \mathcal{L}\{D^{1}\mathbf{f}\} (s)=  \mathcal{L}\{ \mathbf{f'}\} (s)  + f(+0) \mathcal{L}\{D^{1}\mathbf{\delta}\} (s) =   \mathcal{L}\{ \mathbf{f'}\} (s)  + f(+0)  =s  \mathcal{L}\{ \mathbf{f}\} (s)   $$
 
y encontramos

$$\mathcal{L}\{ \mathbf{f'}\} (s)    =s  \mathcal{L}\{ \mathbf{f}\} (s) -  f(+0)$$
 
dado que $\mathbf{f}$ y $\mathbf{f'}$ son dos distribuciones regulares, $\mathcal{L}\{ \mathbf{f'}\} (s) = \mathcal{L}\{ f'\} (s)$ y $\mathcal{L}\{ \mathbf{f}\} (s) = \mathcal{L}\{ f\} (s)$ entonces recobramos la propiedad válida para la Transformada de Laplace de funciones.
 
### Ejemplo: Transformada de Laplace de $D^n \mathbf{\delta}$
Para $n=1$ obtenemos

$$   \mathcal{L}\{ D^1 \mathbf{\delta} \} = s \mathcal{L}\{ \mathbf{\delta}\} = s   $$
 
y en el caso general

$$   \mathcal{L}\{ D^n \mathbf{\delta} \} = s \mathcal{L}\{D^{n-1} \mathbf{\delta}\} = s^n   $$
 
En principio, este tipo de distribución es una forma lineal y continua sobre el espacio de funciones de test, las cuales tienen soporte compacto. Sin embargo, dado que la distribución en sí misma tiene soporte compacto, uno puede ignorar las funciones de test fuera del soporte de la distribución. Esto permite extender la dualidad y considerar las distribuciones con soporte compacto como las formas lineales y continuas sobre el espacio de funciones $C^{\infty}(\mathbb{R};\mathbb{C})$ 

 

$$  \langle \mathbf{T}, \varphi \rangle := \langle \mathbf{T}, \chi \varphi \rangle  $$

donde $\chi \in \mathcal{D}(\mathbb{R};\mathbb{C})$ tal que $\chi=1$ es una vecindad abierta de $\text{supp}(\mathbf{T})$. Esta definición no depende de como se escoja $\chi$.

## Exponenciales como autofunciones de los sistemas LTI

¿Por qué nos interesa descomponer señales en exponenciales complejas? ¿Por qué estas funciones son tan importantes para el análisis de los sistemas LTI?

La respuesta es bastante simple. Las exponenciales son [autofunciones de los sistemas LTI](https://en.wikipedia.org/wiki/Linear_time-invariant_system#Exponentials_as_eigenfunctions). ¿Qué significa esto? Básicamente que cuando una exponencial es la entrada a un sistema LTI, la salida del sistema es la misma exponencial, multiplicada por un factor, denominado autovalor. 

En efecto, consideremos un sistema LTI especificado en el dominio del tiempo por una respuesta al impulso $h(t)$ y supongamos que la entrada es $f(t) = e^{st}$, con $s \in \mathbb{C}$. Sabemos que en un sistema LTI, la salida $g(t)$ la podemos obtener mediante la convolución, es decir: 


$$ g(t) = (h \ast f)(t) = \int_{-\infty}^{\infty} f(t-\tau) h(\tau) d\tau $$

Pero en este caso específico, tenemos que

$$ g(t) = \int_{-\infty}^{\infty} e^{s(t-\tau)} h(\tau) d\tau = e^{st} \int_{-\infty}^{\infty} h(\tau) e^{- s\tau} d\tau = e^{st} H(s)$$

<font color="red">¡¡Reconocemos en el lado derecho la Transformada bi-lateral de Laplace $H(s)$ de la respuesta al impulso $h(t)$!!</font>

Es decir, 

$$ \boxed{g(t) = H(s)e^{st}} $$

como $H(s)$ no depende de la variable independiente $t$ es simplemente un factor de amplificación.


### Exponenciales complejas como autofunciones
Por supuesto, en el caso en que $s = i\omega = i 2 \pi u$, es decir en el dominio de Fourier, las bases de Fourier también son autofunciones de los sistemas LTI, es decir, si la entrada a un sistema LTI es $f(t) = e^{i\omega} = e^{i 2 \pi u}$, la salida será simplemente:

$$ \boxed{g(t) = H(i\omega)e^{i\omega} = H(u)e^{i 2 \pi u t}} $$

<div class="alert alert-block alert-info">
¡La salida de un sistema LTI a una entrada exponencial es simplemente la entrada $e^{st}$ multiplicada por el autovalor $H(s)$! $H(s)$ se conoce como la <a href="https://en.wikipedia.org/wiki/Transfer_function">función de transferencia</a> de un sistema. Conocida esta función $H(s)$, para entradas del tipo exponenciales basta multiplicar la exponencial por el valor de la función de transferencia para encontrar la salida. En el dominio de Fourier $H(u)$, se conoce como la <a href="https://en.wikipedia.org/wiki/Frequency_response">respuesta de frecuencia</a> de un sistema. 
  
<b>En resumen, en ambos casos, la salida es sólo una versión escalada de la entrada.</b>
</div>

## Función de Transferencia

Sabemos además por la propiedad de la convolución de la Transformada bi-lateral de Laplace, que la salida, en el dominio de Laplace, se puede encontrar mediante la relación

$$\boxed{G(s) = H(s)F(s)}$$ 

donde $H(s)$ es la función de transferencia, es decir la Transformada de Laplace de la respuesta al impulso. Esta idea se sintetiza en la siguiente imagen:

## Respuesta de frecuencia

Al evaluar la función de transferencia en $s = i\omega = i 2 \pi u$, obtenemos la respuesta de frecuencia:

$$\boxed{G(i\omega) = H(i\omega)F(i\omega)}$$ 

o bien
$$\boxed{G(u) = H(u)F(u)}$$ 

donde $H(i\omega) = H(u)$ es la respuesta de frecuencia, es decir la Transformada de Fourier de la respuesta al impulso. Esta idea se sintetiza en la siguiente imagen:

<img src="images/Fourier_Transfer_Block.png">

## Representación en magnitud y fase

Tanto la función de transferencia como la respuesta de frecuencia se pueden representar con un magnitud y una fase:

$$ H(s) = |H(s)| e^{i\angle H(s)} $$

$$ H(i\omega) = |H(i\omega)| e^{i\angle H(i\omega)} $$

$$ H(u) = |H(u)| e^{i\angle H(u)} $$

## Evaluación geométrica de la función de transferencia

Para ilustrar este procedimiento, consideremos una función de transferencia $F(s)$, una coordenada $a$ en el eje real y un valor complejo cualquiera $s_1$ en el plano-$s$, como lo indica la figura

<img width="300" src="images/s1.png">

Nos interesa determinar la influencia de la función $F(s)$ con respecto al punto $s_1$, en función de $a$.

***
### Función de transferencia de un cero

Consideremos primero el caso 

$$ F(s) = s - a $$

En este caso, $a$ representa un *cero* de la función $F(s)$. Al evaluar esta funcion en $s_1$ tenemos

$$ F(s_1) = s_1 - a $$

Nos damos cuenta que el valor de la función está determinado por el valor de la magnitud del vector $s_1 - a$ de la figura anterior. Es decir, para este sistema, el valor de la función $F(s_1)$ es directamente proporcional a la magnitud de dicho vector. El ángulo, o fase, de $F(s_1)$, es además igual al ángulo del vector $s_1 - a$ respecto al eje real.

***
### Función de transferencia de un polo

Consideremos ahora el caso 

$$ F(s) = \frac{1}{s - a}$$

En este caso, $a$ representa un *polo* de la función $F(s)$. Al evaluar esta funcion en $s_1$ tenemos

$$ F(s_1) = \frac{1}{s_1 - a}$$

Nos damos cuenta que el valor de la función está determinado por el valor inverso de la magnitud del vector $s_1 - a$ de la figura anterior. Es decir, para este sistema, el valor de la función $F(s_1)$ es inversamente proporcional a la magnitud de dicho vector. El ángulo, o fase, de $F(s_1)$, es además igual al ángulo con signo negativo del vector $s_1 - a$ respecto al eje real.

### Función de transferencia genérica

Considermos ahora una función de transferencia genérica dada por:

\begin{equation}
F(s) =  K \cdot \frac{\prod_{\mu = 0}^{Q}(s - s_{0 \mu})}{\prod_{\nu = 0}^{P}(s - s_{\infty \nu})}
\end{equation}

<div class="alert alert-block alert-info">
Siguiendo el razonamiento anterior podemos establecer que:
<ul>
<li>$F(s_1)$ tiene una magnitud directamente proporcional a la multiplicación de los vectores cero e inversamente proporcional a multiplicación de los vectores polos.
<li>$F(s_1)$ tiene una fase igual a la suma de los ángulos de los vectores cero menos la suma de los ángulos de los vectores polos.
</ul>

## Evaluación geométrica de la respuesta de frecuencia

Cuando evaluamos la función de transferencia en $s = i\omega$, estamos restringiendo el ámbito de evaluación el eje imaginario, es decir a la Transformada de Fourier. Esto implica que el lugar geométrico de interés para aplicar la interpretación geométrica anteriormente descrita es el eje imaginario, tal como lo indica la siguiente animación.

### Filtros selectivos de frecuencia
El posicionar polos y ceros en el plano-$s$ permite determinar curvas de respuesta de frecuencia específicas. Estas curvas implican que ciertas frecuencias se atenuarán mientras otras se enfatizarán. Por lo tanto una función de transferencia es básicamente un **filtro selectivo en frecuencia**, cuando se evalúa en el eje imaginario.

En un filtro, el *orden* está determinado por el número máximo de polos o ceros que tenga la función de transferencia.

***
### Ejemplo: Filtro pasa bajos de primer orden

Veamos el caso de un filtro pasa bajos de primer orden con función de transferencia

$$ F(s) = \frac{1}{s+a} $$

Lo que este filtro hace es enfatizar la frecuencia cero dependiendo del valor de $a$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sympy.plotting.plot import MatplotlibBackend
import sympy as sym
sym.init_printing()

s = sym.symbols('s', complex=True)
u, a = sym.symbols('u a', real=True)
H = 1/(s + a)

L = 1
a1 = -0.1
a2 = -0.5
a3 = -2


p1 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a1), (u, -L, L), label='a=1', xlabel='$u$', ylabel='$ | H(u) | $', line_color='r', show=False);
p2 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a2), (u, -L, L), label='b=2', xlabel='$u$', ylabel='$ | H(u) | $', line_color='g', show=False);
p3 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a3), (u, -L, L), label='c=3', xlabel='$u$', ylabel='$ | H(u) | $', line_color='b', show=False);
p1.extend(p2)
p1.extend(p3)

backend = MatplotlibBackend(p1)
backend.process_series()
backend.ax.spines['left'].set_position(('axes', 0))
backend.ax.spines['bottom'].set_position(('axes', 0))
backend.ax.legend(['$a$ = %.1f' %(a1),'$a$ = %.1f' %(a2),'$a$ = %.1f' %(a3),])

plt.xscale('linear')
plt.grid(which='none')
plt.title("Filtro pasa bajos de primer orden: F(s) = " + str(H))



pp1 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a1), (u, -L, L), label='a=1', xlabel='$u$', ylabel='$< H(u) $', line_color='r', show=False);
pp2 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a2), (u, -L, L), label='b=2', xlabel='$u$', ylabel='$< H(u) $', line_color='g', show=False);
pp3 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,a3), (u, -L, L), label='c=3', xlabel='$u$', ylabel='$< H(u) $', line_color='b', show=False);
pp1.extend(pp2)
pp1.extend(pp3)

backend = MatplotlibBackend(pp1)
backend.process_series()
backend.ax.spines['left'].set_position(('axes', 0))
backend.ax.spines['bottom'].set_position(('axes', 0))
backend.ax.legend(['$a$ = %.1f' %(a1),'$a$ = %.1f' %(a2),'$a$ = %.1f' %(a3),])

plt.xscale('linear')
plt.grid(which='none')
plt.title("Filtro pasa bajos de primer orden: F(s) = " + str(H))
plt.show()
```

### Ejemplo: filtro rechaza banda de segundo orden 

Veamos el caso de un filtro pasa altos de segundo orden con función de transferencia

$$ F(s) = (s+c)(s+\bar{c}) $$

con $c \in \mathbb{C}$. Lo que este filtro hace es atenuar la frecuencia más cercana a la posición de los polos determinados por $c$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sympy.plotting.plot import MatplotlibBackend
import sympy as sym
sym.init_printing()

s, a, b = sym.symbols('s a b', complex=True)
u = sym.symbols('u', real=True)
H = ((s+a)*(s+b))

L = .25
R = -0.01
c1 = R + 0.2*sym.I
c2 = R + 0.5*sym.I
c3 = R + sym.I
cc1 = R - 0.2*sym.I
cc2 = R - 0.5*sym.I
cc3 = R - sym.I

p1 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c1).subs(b,cc1), (u, -L, L), label='a=1', xlabel='$u$', ylabel='$ | H(u) | $', line_color='r', show=False);
p2 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c2).subs(b,cc2), (u, -L, L), label='b=2', xlabel='$u$', ylabel='$ | H(u) | $', line_color='g', show=False);
p3 = sym.plot(sym.Abs(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c3).subs(b,cc3), (u, -L, L), label='c=3', xlabel='$u$', ylabel='$ | H(u) | $', line_color='b', show=False);
p1.extend(p2)
p1.extend(p3)

backend = MatplotlibBackend(p1)
backend.process_series()
backend.ax.spines['left'].set_position(('axes', 0))
backend.ax.spines['bottom'].set_position(('axes', 0))
backend.ax.legend(['$a,b$ = %s , %s' %(c1,cc1),'$a,b$ = %s , %s' %(c2,cc2),'$a,b$ = %s , %s' %(c3,cc3),])

plt.xscale('linear')
plt.grid(which='none')
plt.title("Filtro rechaza banda de segundo orden: F(s) = " + str(H))



pp1 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c1).subs(b,cc1), (u, -L, L), label='a=1', xlabel='$u$', ylabel='$< H(u) $', line_color='r', show=False);
pp2 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c2).subs(b,cc2), (u, -L, L), label='b=2', xlabel='$u$', ylabel='$< H(u) $', line_color='g', show=False);
pp3 = sym.plot(sym.arg(H).subs(s, 2*sym.pi*u*sym.I).subs(a,c3).subs(b,cc3), (u, -L, L), label='c=3', xlabel='$u$', ylabel='$< H(u) $', line_color='b', show=False);
pp1.extend(pp2)
pp1.extend(pp3)

backend = MatplotlibBackend(pp1)
backend.process_series()
backend.ax.spines['left'].set_position(('axes', 0))
backend.ax.spines['bottom'].set_position(('axes', 0))
backend.ax.legend(['$a,b$ = %s , %s' %(c1,cc1),'$a,b$ = %s , %s' %(c2,cc2),'$a,b$ = %s , %s' %(c3,cc3),])

plt.xscale('linear')
plt.grid(which='none')
plt.title("Filtro rechaza banda de segundo orden: F(s) = " + str(H))
plt.show()
```

## Filtro pasa bajos ideal

El filtro pasabajos ideal es aquel que multiplica, en frecuencia, la Transformada de Fourier de la señal de entrada por un rect. De esta manera, todas las bajas frecuencias contenidas en el rect se multiplican por 1 y las que no, se anulan al ser multiplicadas por cero. La siguiente figura ilustra este filtro:

<img width="600" src="images/idealLP.png">

Esto implica que, en el dominio del tiempo, un filtro pasa bajos ideal es simplemente convolucionar con una señal sinc, graficada en la figura más abajo.

<img width="600" src="images/sinc.png">


Es importante notar que la respuesta de fase del filtro ideal debe ser lineal. Esto implica que el desfase de todas las componentes sinusoidales contenidas en la entrada se desfasan en una misma proporción respecto a su frecuencia. De otra manera el desfase no sería parejo para todas las frecuencias y la salida estaría distorsionada. La alternativa de desfase cero no es práctica, por que el cálculo de la convolución con el sinc toma tiempo y necesariamente tiene que haber un tiempo de desfase.
